## 섹션 2: UV 패키지 매니저

### UV란 무엇인가요?

Python 패키지를 설치할 때 보통 `pip`을 사용하죠? **UV는 `pip`의 업그레이드 버전**입니다.

**자전거 vs 스포츠카 비유**
- `pip` = 자전거: 느리지만 어디서나 있고, 익숙하다
- `UV` = 스포츠카: Rust 언어로 만들어져 pip보다 **10~100배 빠르다!**

### pip vs UV 비교

| 항목 | pip | UV |
|------|-----|----|
| 속도 | 보통 | 10~100배 빠름 |
| 언어 | Python | Rust |
| 가상환경 관리 | 별도 venv 필요 | 자동 통합 |
| 프로젝트 관리 | requirements.txt | pyproject.toml |

> 이 커리큘럼은 UV 기반으로 설계되어 있으며, Jupyter 환경에서도 `!uv pip install`을 사용합니다.

````python
# UV 설치 여부 확인
import subprocess
import sys

result = subprocess.run(['uv', '--version'], capture_output=True, text=True)
if result.returncode == 0:
    print(f'[OK] UV가 설치되어 있습니다: {result.stdout.strip()}')
else:
    print('[FAIL] UV가 설치되어 있지 않습니다.')
    print('아래 섹션의 설치 방법을 참고하세요.')
````

````python
import importlib.metadata

print('=== 패키지 설치 확인 ===')
print()

# 확인할 모듈 이름과 실제 패키지 이름 매핑
packages = [
    ("langchain", "langchain"),
    ("langchain_core", "langchain-core"),
    ("langchain_community", "langchain-community"),
    ("langchain_google_genai", "langchain-google-genai"),
    ("dotenv", "python-dotenv")
]

for module_name, package_name in packages:
    try:
        # 모듈이 정상적으로 import 되는지 확인
        __import__(module_name)
        # 패키지 메타데이터에서 버전을 안전하게 추출
        version = importlib.metadata.version(package_name)
        print(f'[OK] {package_name}: {version}')
        
    except ImportError:
        print(f'[FAIL] {package_name}: 설치 실패 (모듈을 가져올 수 없습니다)')
    except importlib.metadata.PackageNotFoundError:
        print(f'[FAIL] {package_name}: 패키지를 찾을 수 없습니다')

print()
print('모두 [OK]라면 패키지 설치 완료!')
````

## 섹션 4: Gemini API 키

### Gemini API란?

**API (Application Programming Interface)**는 서비스를 코드로 사용할 수 있게 해주는 "창구"입니다.

**레스토랑 비유**
- 구글 AI 서버 = 주방
- Gemini API = 주문을 받는 웨이터
- API 키 = 내 회원권 (없으면 주문 불가!)
- 우리 코드 = 손님

```
우리 코드  →  [API 키]  →  Gemini API  →  구글 AI 서버
   손님    →  회원권   →     웨이터    →       주방
```

### 왜 LLM API를 쓰나요?

- **직접 모델 실행은 불가능**: GPT, Gemini 같은 대형 모델은 수백 GB라 내 컴퓨터에서 실행 불가
- **클라우드 활용**: 구글이 강력한 서버에서 모델을 실행하고, 우리는 결과만 받음
- **Gemini는 무료 티어 제공**: Google AI Studio에서 API 키 발급 시 무료로 사용 가능

### Gemini API 키 발급 방법 (단계별 안내)

**Step 1** 브라우저를 열고 [Google AI Studio](https://aistudio.google.com/) 접속

**Step 2** 구글 계정으로 로그인

**Step 3** 왼쪽 사이드바에서 **"Get API key"** 클릭

**Step 4** **"Create API key"** 버튼 클릭

**Step 5** 생성된 API 키를 **복사** (예: `AIzaSy...` 형태의 긴 문자열)

**Step 6** 아래 코드 셀로 이동해서 API 키 입력

> **주의**: API 키는 절대로 다른 사람에게 공유하지 마세요! 악용되면 요금이 발생할 수 있습니다.

> **참고**: Gemini API는 일정량 무료로 제공됩니다. 학습 목적으로 사용하기에 충분합니다.

````python
# API 키를 안전하게 입력받기
# getpass를 사용하면 입력한 내용이 화면에 보이지 않습니다 (비밀번호 입력처럼!)
import getpass

print('Gemini API 키를 입력하세요:')
print('(입력 중에는 화면에 아무것도 표시되지 않습니다 - 정상입니다!)')
print()

api_key = getpass.getpass('API 키 입력: ')

if api_key:
    masked = api_key[:8] + '*' * (len(api_key) - 8)
    print(f'\n[OK] API 키 입력 완료! ({masked})')
    print('다음 셀로 이동해서 .env 파일을 생성하세요.')
else:
    print('\n[FAIL] API 키가 입력되지 않았습니다. 다시 실행해주세요.')
````

## 섹션 5: .env 파일 설정

### `.env` 파일이란?

**일기장 비유**
- 일기에 비밀을 적어두듯, `.env` 파일에 비밀 정보(API 키, 비밀번호)를 저장합니다
- 코드 파일(`*.py`, `*.ipynb`)은 GitHub에 올려도 되지만, `.env`는 절대 올리면 안 됩니다!

### 왜 코드에 직접 쓰면 안 될까요?

```python
# 이렇게 하면 절대 안 됩니다!
api_key = "AIzaSyABCDEF12345678"  # GitHub에 올리면 전 세계가 봅니다
```

```python
# 이렇게 해야 합니다!
import os
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")  # .env 파일에서 안전하게 읽기
```

### `.gitignore`란?

`.gitignore` 파일에 `.env`를 추가하면, Git이 자동으로 이 파일을 무시합니다.
이 프로젝트의 `.gitignore`에 이미 `.env`가 포함되어 있습니다.

````python
# .env 파일 자동 생성
# 위에서 입력한 API 키를 .env 파일에 저장합니다
import os

# 이 노트북은 00_setup/ 폴더에 있으므로, 상위 폴더(프로젝트 루트)에 .env를 저장합니다
notebook_dir = os.path.dirname(os.path.abspath('__file__'))
project_root = os.path.dirname(notebook_dir)
env_path = os.path.join(project_root, '.env')

env_content = f"""# Google Gemini API 키
# 발급: https://aistudio.google.com/
GOOGLE_API_KEY={api_key}
"""

with open(env_path, 'w', encoding='utf-8') as f:
    f.write(env_content)

print(f'[OK] .env 파일 생성 완료!')
print(f'저장 위치: {env_path}')
print()
print('주의: 이 파일은 절대 GitHub에 올리지 마세요!')
print('      (.gitignore에 이미 추가되어 있어서 자동으로 보호됩니다)')
````

````python
# .env 파일 불러오기 및 API 키 확인
import os
from dotenv import load_dotenv

notebook_dir = os.path.dirname(os.path.abspath('__file__'))
project_root = os.path.dirname(notebook_dir)
env_path = os.path.join(project_root, '.env')

loaded = load_dotenv(env_path)

print('=== .env 파일 로드 결과 ===')
if loaded:
    print('[OK] .env 파일 로드 성공!')
else:
    print('[주의] .env 파일을 찾지 못했습니다. 경로를 확인하세요.')

print()
print('=== API 키 확인 ===')
loaded_key = os.getenv('GOOGLE_API_KEY')
if loaded_key and loaded_key != 'your_api_key_here':
    masked = loaded_key[:10] + '...' + loaded_key[-4:]
    print(f'[OK] GOOGLE_API_KEY: {masked}')
    print('API 키가 정상적으로 로드되었습니다!')
else:
    print('[FAIL] API 키가 없거나 아직 설정되지 않았습니다.')
    print('위 셀들을 순서대로 다시 실행해주세요.')
````

## 섹션 6: 환경 검증

### 최종 점검 방법

모든 설정이 끝났습니다! 이제 마지막으로 전체 환경이 올바르게 설정되었는지 확인합니다.

아래 두 개의 셀을 실행하면:

1. **종합 체크** - 각 항목별로 [OK] 또는 [FAIL]로 상태를 보여줍니다
2. **모델 import 테스트** - LangChain에서 Gemini 모델을 불러올 수 있는지 확인합니다

모든 항목이 [OK]라면 **모듈 01로 넘어갈 준비가 완료**된 것입니다!

````python
# 종합 환경 체크
import os
import sys
import importlib.metadata  # 버전 확인을 위해 추가
from dotenv import load_dotenv

def check_all():
    results = []

    # 1. Python 버전 확인
    v = sys.version_info
    if v >= (3, 10):
        results.append(('Python 버전', True, f'{v.major}.{v.minor}.{v.micro}'))
    else:
        results.append(('Python 버전', False, f'{v.major}.{v.minor} (3.10 이상 권장)'))

    # 2. langchain 확인
    try:
        import langchain
        version = importlib.metadata.version("langchain")
        results.append(('langchain', True, version))
    except ImportError:
        results.append(('langchain', False, '미설치'))
    except importlib.metadata.PackageNotFoundError:
        results.append(('langchain', False, '패키지 찾을 수 없음'))

    # 3. langchain-google-genai 확인
    try:
        import langchain_google_genai
        version = importlib.metadata.version("langchain-google-genai")
        results.append(('langchain-google-genai', True, version))
    except ImportError:
        results.append(('langchain-google-genai', False, '미설치'))
    except importlib.metadata.PackageNotFoundError:
        results.append(('langchain-google-genai', False, '패키지 찾을 수 없음'))

    # 4. python-dotenv 확인
    try:
        import dotenv
        version = importlib.metadata.version("python-dotenv")
        results.append(('python-dotenv', True, version))
    except ImportError:
        results.append(('python-dotenv', False, '미설치'))
    except importlib.metadata.PackageNotFoundError:
        results.append(('python-dotenv', False, '패키지 찾을 수 없음'))

    # 5. .env 파일 확인
    notebook_dir = os.path.dirname(os.path.abspath('__file__'))
    project_root = os.path.dirname(notebook_dir)
    env_path = os.path.join(project_root, '.env')
    if os.path.exists(env_path):
        results.append(('.env 파일', True, f'존재 ({env_path})'))
    else:
        results.append(('.env 파일', False, '파일 없음 - 위 셀을 실행하세요'))

    # 6. API 키 확인
    load_dotenv(env_path)
    api_key = os.getenv('GOOGLE_API_KEY')
    if api_key and api_key != 'your_api_key_here' and len(api_key) > 10:
        masked = api_key[:10] + '...'
        results.append(('GOOGLE_API_KEY', True, masked))
    else:
        results.append(('GOOGLE_API_KEY', False, '미설정 또는 예시값'))

    # 결과 출력
    print('=' * 50)
    print('       환경 설정 종합 체크 결과')
    print('=' * 50)
    all_ok = True
    for name, ok, detail in results:
        status = '[OK]  ' if ok else '[FAIL]'
        print(f'{status} {name:<25} {detail}')
        if not ok:
            all_ok = False

    print('=' * 50)
    if all_ok:
        print('모든 항목 통과! 모듈 01로 넘어갈 준비가 됐습니다!')
    else:
        print('[FAIL] 항목을 확인하고 해당 셀을 다시 실행하세요.')

check_all()
````

````python
# Gemini 모델 import 테스트
# 실제 API를 호출하지 않고, 모델 클래스를 불러올 수 있는지만 확인합니다
print('=== Gemini 모델 Import 테스트 ===')
print()

try:
    from langchain_google_genai import ChatGoogleGenerativeAI
    print('[OK] ChatGoogleGenerativeAI import 성공!')
    print()
    print('다음 모듈에서 이 클래스로 실제 AI와 대화합니다:')
    print()
    print('  from langchain_google_genai import ChatGoogleGenerativeAI')
    print('  llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")')
    print('  response = llm.invoke("안녕하세요!")')
    print()
    print('모듈 00 환경 설정이 완료되었습니다!')
except ImportError as e:
    print(f'[FAIL] Import 실패: {e}')
    print()
    print('아래 명령어로 패키지를 다시 설치해보세요:')
    print('  !uv pip install langchain-google-genai')
````
