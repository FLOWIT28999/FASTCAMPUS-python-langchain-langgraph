## 이 챕터에서 배울 내용

- 딕셔너리(`dict`)로 키-값 쌍 데이터를 저장하고 조회한다
- 집합(`set`)으로 중복을 제거하고 합집합·교집합·차집합 연산을 수행한다

- 인덱싱과 슬라이싱으로 원하는 값을 정확히 꺼낼 수 있다
- 이름(키)으로 값을 저장하고 빠르게 조회할 수 있다
- 중복 데이터를 집합으로 깔끔하게 정리할 수 있다
- 리스트, 튜플, 딕셔너리, 집합 중 어떤 것을 써야 할지 판단할 수 있다

## 1. 리스트 (list)

리스트는 **순서가 있고 수정 가능한(mutable) 컨테이너**입니다. 여러 값을 하나의 변수에 담을 수 있으며, 서로 다른 자료형을 혼합할 수도 있습니다.

```python
fruits = ["사과", "바나나", "포도"]    # 문자열 리스트
numbers = [1, 2, 3, 4, 5]             # 정수 리스트
mixed  = ["파이썬", 3.14, True, None]  # 혼합 리스트
```

### 인덱싱 — 위치로 값 꺼내기

파이썬에서 인덱스는 **0부터** 시작합니다. 음수 인덱스는 **뒤에서부터** 셉니다.

| 리스트 | `"사과"` | `"바나나"` | `"포도"` | `"딸기"` | `"수박"` |
|--------|----------|-----------|---------|---------|--------|
| 양수 인덱스 | `0` | `1` | `2` | `3` | `4` |
| 음수 인덱스 | `-5` | `-4` | `-3` | `-2` | `-1` |

### 슬라이싱 — 범위로 잘라내기

```python
리스트[start:stop:step]
```

- `start`: 시작 인덱스 (포함, 생략 시 처음부터)
- `stop`: 끝 인덱스 (**미포함**, 생략 시 끝까지)
- `step`: 간격 (생략 시 1)

### 주요 메서드

| 메서드 | 설명 | 예시 |
|--------|------|------|
| `append(값)` | 맨 뒤에 추가 | `lst.append(9)` |
| `insert(i, 값)` | 인덱스 i에 삽입 | `lst.insert(0, 0)` |
| `remove(값)` | 첫 번째로 나오는 값 제거 | `lst.remove(3)` |
| `pop(i)` | 인덱스 i의 값 꺼내기 (생략 시 마지막) | `lst.pop()` |
| `sort()` | 오름차순 정렬 (원본 변경) | `lst.sort()` |
| `reverse()` | 순서 뒤집기 (원본 변경) | `lst.reverse()` |
| `len(lst)` | 요소 개수 반환 (내장 함수) | `len(lst)` |

````python
# 리스트 생성과 인덱싱·슬라이싱
fruits = ["사과", "바나나", "포도", "딸기", "수박"]
print(fruits)
print(fruits[0])    # 첫 번째: 사과
print(fruits[-1])   # 마지막: 수박
print(fruits[1:3])  # 슬라이싱: ['바나나', '포도']
print(fruits[::2])  # 홀수 인덱스: ['사과', '포도', '수박']
````

````python
# 리스트 추가·삭제 메서드
numbers = [3, 1, 4, 1, 5, 9]
numbers.append(2)       # 맨 뒤에 추가
numbers.insert(0, 0)    # 인덱스 0에 삽입
numbers.remove(1)       # 값 1 중 첫 번째 제거
popped = numbers.pop()  # 마지막 요소 꺼내기
print(numbers)
print("꺼낸 값:", popped)
````

````python
# 정렬과 길이
scores = [85, 92, 78, 95, 88]
print("길이:", len(scores))
scores.sort()       # 오름차순 정렬 (원본 변경)
print("정렬:", scores)
scores.reverse()    # 역순 (원본 변경)
print("역순:", scores)
````

## 2. 튜플 (tuple)

튜플은 **순서가 있지만 수정 불가(immutable)**한 컨테이너입니다. 한번 만든 튜플은 요소를 추가·변경·삭제할 수 없습니다.

```python
point = (3, 7)          # 괄호로 생성
rgb   = 255, 128, 0     # 괄호 없이도 생성 가능
```

### 주의: 요소가 1개인 튜플

요소가 1개일 때는 반드시 **쉼표**를 붙여야 튜플로 인식합니다.

```python
single     = (42,)  # tuple — 쉼표 있음
not_tuple  = (42)   # int   — 쉼표 없음, 그냥 괄호
```

### 리스트 vs 튜플

| | 리스트 `list` | 튜플 `tuple` |
|---|---|---|
| 수정 가능 여부 | 가능 (mutable) | 불가 (immutable) |
| 생성 문법 | `[...]` | `(...)` |
| 주요 용도 | 항목이 바뀌는 목록 | 바뀌면 안 되는 데이터 |

### 튜플 언패킹 (Unpacking)

튜플의 값을 여러 변수에 **한 번에 할당**하는 기법입니다.

```python
x, y = (3, 7)   # x = 3, y = 7
```

````python
# 튜플 생성과 불변성
point = (3, 7)
rgb = (255, 128, 0)
print(point[0], point[1])

# 불변성: 아래 줄 주석 해제 시 TypeError 발생
# point[0] = 10

# 요소 1개짜리 튜플은 쉼표 필수
single    = (42,)
not_tuple = (42)
print(type(single))    # <class 'tuple'>
print(type(not_tuple)) # <class 'int'>
````

````python
# 튜플 언패킹
point = (3, 7)
x, y = point
print("x:", x, "y:", y)

# 여러 변수 동시 할당
name, age, city = ("김파이썬", 25, "서울")
print(name, age, city)

# swap (값 교환) — 튜플 언패킹 활용
a, b = 10, 20
a, b = b, a
print(a, b)  # 20 10
````

## 3. 딕셔너리 (dict)

딕셔너리는 **키(key)와 값(value)의 쌍으로 데이터를 저장**하는 자료구조입니다. 순서 있는 목록(리스트)과 달리, 이름(키)으로 값을 조회합니다.

```python
person = {"name": "김파이썬", "age": 25}
#          키       값          키     값
```

- 키는 **변경 불가 타입**(문자열, 정수 등)만 사용할 수 있습니다. 리스트는 키로 사용 불가.
- Python 3.7 이후부터 삽입 순서가 유지됩니다.

### `dict[key]` vs `dict.get(key)` 차이

| 방법 | 키가 존재할 때 | 키가 없을 때 |
|------|-------------|-------------|
| `dict["key"]` | 값 반환 | `KeyError` 발생 |
| `dict.get("key")` | 값 반환 | `None` 반환 |
| `dict.get("key", 기본값)` | 값 반환 | 기본값 반환 |

### 주요 메서드

| 메서드 | 설명 |
|--------|------|
| `keys()` | 모든 키를 반환 |
| `values()` | 모든 값을 반환 |
| `items()` | (키, 값) 쌍 목록 반환 |
| `get(키, 기본값)` | 키로 값 조회, 없으면 기본값 |
| `update({...})` | 다른 딕셔너리로 일괄 업데이트 |
| `pop(키)` | 키 제거 후 값 반환 |

````python
# 딕셔너리 생성과 값 접근
person = {
    "name": "김파이썬",
    "age": 25,
    "city": "서울"
}
print(person["name"])                    # 키로 직접 접근
print(person.get("age"))                 # get()으로 접근
print(person.get("email"))     # 키 없을 때 기본값 반환
````

````python
# 딕셔너리 추가·수정과 메서드
person = {"name": "김파이썬", "age": 25}
person["city"] = "서울"   # 새 키 추가
person["age"] = 26        # 기존 키 수정
print(person)

print(list(person.keys()))    # 키 목록
print(list(person.values()))  # 값 목록
print(list(person.items()))   # (키, 값) 쌍 목록
````

````python
# 키 삭제와 포함 여부 확인
info = {"a": 1, "b": 2, "c": 3}
removed = info.pop("b")     # 키 "b" 제거 후 값 반환
print(info)
print("제거된 값:", removed)

# in 연산자로 키 존재 여부 확인
print("a" in info)   # True
print("b" in info)   # False
````

## 4. 집합 (set)

집합은 **중복 없음, 순서 없음**의 특성을 가진 자료구조입니다. 수학의 집합과 동일한 개념입니다.

```python
s = {1, 2, 3}          # 중괄호로 생성
s = set([1, 2, 2, 3])  # 리스트로부터 생성 (중복 제거됨)
```

### 주의: 빈 집합 생성

`{}` 단독으로 쓰면 **빈 딕셔너리**가 만들어집니다. 빈 집합은 반드시 `set()`을 사용하세요.

```python
empty_dict = {}      # dict — 빈 딕셔너리
empty_set  = set()   # set  — 빈 집합
```

### 집합 연산

| 연산자 | 메서드 | 의미 | 설명 |
|--------|--------|------|------|
| `A \| B` | `A.union(B)` | 합집합 | A 또는 B에 있는 모든 요소 |
| `A & B` | `A.intersection(B)` | 교집합 | A와 B 모두에 있는 요소 |
| `A - B` | `A.difference(B)` | 차집합 | A에는 있지만 B에는 없는 요소 |

````python
# 집합 생성과 중복 제거
# 중복이 자동으로 제거됨
numbers = {3, 1, 4, 1, 5, 9, 2, 6, 5}
print(numbers)  # 순서 없음, 중복 없음
````

````python

# 리스트의 중복을 제거할 때 활용
raw = [1, 2, 2, 3, 3, 3, 4]
unique = list(set(raw))
print(unique)
````

````python

# 빈 집합은 set()으로 생성
empty_dict = {}       # dict
empty_set  = set()    # set
print(type(empty_dict))  # <class 'dict'>
print(type(empty_set))   # <class 'set'>
````

````python
# 집합 연산
python_students = {"김철수", "이영희", "박민준"}
java_students   = {"이영희", "최수진", "박민준"}

print("합집합 (전체):", python_students | java_students)
print("교집합 (둘 다 수강):", python_students & java_students)
print("차집합 (파이썬만 수강):", python_students - java_students)
````

## 5. 자료구조 비교 정리

4가지 자료구조의 핵심 특성을 비교합니다.

| 자료구조 | 순서 | 수정 가능 | 중복 허용 | 대표 용도 |
|----------|------|-----------|-----------|----------|
| `list` | 있음 | 가능 | 허용 | 순서 있는 항목 목록 |
| `tuple` | 있음 | 불가 | 허용 | 변경되면 안 되는 데이터 |
| `dict` | (3.7+ 삽입순) | 가능 | 키는 불가 | 키-값 매핑 |
| `set` | 없음 | 가능 | 불가 | 중복 제거, 집합 연산 |

### 어떤 자료구조를 선택할까?

- 순서 있는 목록이 필요하고, 나중에 내용이 바뀔 수 있다 → **`list`**
- 한번 정하면 절대 바뀌면 안 되는 데이터(좌표, 색상값 등) → **`tuple`**
- 이름(키)으로 빠르게 조회해야 한다(학생 정보, 설정값 등) → **`dict`**
- 중복을 없애거나 두 그룹의 교집합·차집합이 필요하다 → **`set`**

````python
# 같은 데이터를 4가지 방식으로

# 리스트: 순서 있음, 수정 가능
scores_list = [85, 92, 78]

# 튜플: 순서 있음, 수정 불가
scores_tuple = (85, 92, 78)

# 집합: 중복 제거, 순서 없음
scores_set = {85, 92, 78}

# 딕셔너리: 키-값 쌍
scores_dict = {"국어": 85, "영어": 92, "수학": 78}

print(type(scores_list))
print(type(scores_tuple))
print(type(scores_set))
print(type(scores_dict))
````

## 챕터 정리

이번 챕터에서 배운 내용을 정리합니다.

- **리스트 `list`**: `[]`로 생성, 인덱스(0부터)로 접근, 슬라이싱으로 부분 추출. `append`, `insert`, `remove`, `pop`, `sort`, `reverse`로 수정 가능
- **튜플 `tuple`**: `()`로 생성, 수정 불가(immutable). 1개짜리 튜플은 `(42,)` 쉼표 필수. 언패킹으로 여러 변수에 한 번에 할당
- **딕셔너리 `dict`**: `{키: 값}`으로 생성. `dict[키]`는 키 없으면 오류, `dict.get(키)`는 키 없으면 `None`. `keys()`, `values()`, `items()`로 내용 조회
- **집합 `set`**: `{값, ...}` 또는 `set()`으로 생성. 중복 자동 제거, 순서 없음. `|`(합집합), `&`(교집합), `-`(차집합) 연산 가능. 빈 집합은 반드시 `set()`

### 4가지 자료구조 한눈에 비교

| 자료구조 | 생성 문법 | 순서 | 수정 | 중복 | 키-값 쌍 |
|----------|-----------|------|------|------|----------|
| `list` | `[1, 2, 3]` | 있음 | 가능 | 허용 | 아니오 |
| `tuple` | `(1, 2, 3)` | 있음 | 불가 | 허용 | 아니오 |
| `dict` | `{"k": 1}` | (삽입순) | 가능 | 키 불가 | 예 |
| `set` | `{1, 2, 3}` | 없음 | 가능 | 불가 | 아니오 |
