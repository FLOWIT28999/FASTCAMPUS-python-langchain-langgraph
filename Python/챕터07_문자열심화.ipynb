## 1. 인덱싱, 슬라이싱, 불변성

### 1-1. 인덱싱과 슬라이싱 복습

챕터 04에서 배운 시퀀스 인덱싱/슬라이싱은 문자열에도 동일하게 적용된다.

| 표현식 | 의미 |
|---|---|
| `s[i]` | i번째 문자 (음수면 뒤에서부터) |
| `s[start:stop]` | start 이상, stop 미만 구간 |
| `s[start:stop:step]` | step 간격으로 추출 |
| `s[::-1]` | 문자열 전체를 거꾸로 |

### 1-2. 문자열 불변성 (immutable)

문자열은 한 번 만들어지면 개별 문자를 바꿀 수 없다.  
`s[0] = "H"` 처럼 특정 위치에 값을 대입하면 `TypeError`가 발생한다.  
수정이 필요할 때는 **슬라이싱과 연결**로 새 문자열을 만들어야 한다.

```python
# 인덱싱 & 슬라이싱 복습
s = "Hello, Python!"

print(s[0])       # 첫 번째 문자
print(s[-1])      # 마지막 문자
print(s[7:13])    # 7 이상 13 미만 → Python
print(s[::-1])    # 문자열 뒤집기
```

```python
# 불변성 시연
s = "hello"

# s[0] = "H"  # 주석 해제 시 TypeError: 'str' object does not support item assignment

# 수정하려면 슬라이싱으로 새 문자열을 만든다
s2 = "H" + s[1:]
print(s2)   # Hello
print(s)    # hello — 원본은 변하지 않는다
```

## 2. 분리와 합치기: split() / join()

| 메서드 | 역할 | 반환 타입 |
|---|---|---|
| `split(sep)` | 구분자를 기준으로 문자열을 나눈다 | 리스트 |
| `" ".join(iterable)` | 리스트 요소를 구분자로 이어 붙인다 | 문자열 |

- `split()` 에 인자를 생략하면 공백(스페이스, 탭, 줄바꿈)을 기준으로 나눈다
- `join()` 은 **구분자 문자열**에 `.join(리스트)` 형태로 호출한다
- 두 메서드는 쌍으로 이해하면 좋다: split → 처리 → join

```python
# split
sentence = "파이썬 프로그래밍 기초 강좌"
words = sentence.split()          # 공백 기준 분리
print(words)

csv_line = "홍길동,25,서울"
fields = csv_line.split(",")      # 쉼표 기준 분리
print(fields)

print()

# join
print(" / ".join(words))           # 리스트 → 문자열 (슬래시 구분)
print(", ".join(["사과", "바나나", "포도"]))  # 직접 리스트를 넘겨도 된다
```

## 3. 공백 제거, 치환, 대소문자

| 메서드 | 역할 |
|---|---|
| `strip()` | 문자열 앞뒤 공백(또는 지정 문자) 제거 |
| `lstrip()` | 왼쪽(앞쪽)만 제거 |
| `rstrip()` | 오른쪽(뒤쪽)만 제거 |
| `replace(old, new)` | `old`를 모두 찾아 `new`로 치환 |
| `upper()` | 전체를 대문자로 변환 |
| `lower()` | 전체를 소문자로 변환 |

> `repr()` 을 사용하면 문자열 양 끝의 공백이 눈에 보이게 따옴표로 감싸 출력된다.

```python
# strip
raw = "   파이썬   "
print(repr(raw.strip()))    # '파이썬'
print(repr(raw.lstrip()))   # '파이썬   '
print(repr(raw.rstrip()))   # '   파이썬'

print()
```

```python

# replace
text = "Hello, World! Hello, Python!"
print(text.replace("Hello", "안녕"))   # 모든 'Hello' 를 '안녕' 으로 치환

print()
```

```python

# upper / lower
msg = "Python is FUN"
print(msg.upper())    # PYTHON IS FUN
print(msg.lower())    # python is fun
```

## 4. 검색 메서드

| 메서드 | 반환값 | 없거나 False일 때 |
|---|---|---|
| `find(sub)` | 첫 번째 등장 인덱스 (int) | `-1` |
| `count(sub)` | 등장 횟수 (int) | `0` |
| `startswith(prefix)` | True / False | `False` |
| `endswith(suffix)` | True / False | `False` |

- `find()` 는 찾지 못하면 예외 대신 `-1` 을 반환하므로 안전하게 사용할 수 있다
- `startswith()` / `endswith()` 는 파일 확장자, URL 접두사 검사 등에 자주 쓰인다

```python
s = "파이썬 프로그래밍, 파이썬은 재미있다"

print(s.find("파이썬"))        # 0  — 첫 번째 등장 위치
print(s.find("자바"))          # -1 — 없으면 -1
print(s.count("파이썬"))       # 2  — 등장 횟수

print()

filename = "report_2026.pdf"
print(filename.startswith("report"))   # True
print(filename.endswith(".pdf"))       # True
print(filename.endswith(".xlsx"))      # False
```

## 5. 문자열 포매팅

변수를 문자열 안에 삽입하는 방법은 세 가지다.

| 방식 | 예시 | 권장 여부 |
|---|---|---|
| f-string | `f"이름: {name}"` | 권장 (Python 3.6 이상) |
| `format()` | `"이름: {}".format(name)` | 사용 가능 |
| `%` 포매팅 | `"이름: %s" % name` | 구형, 비권장 |

**f-string 서식 지정 문법** — `{값:서식코드}`

| 서식코드 | 의미 | 예 |
|---|---|---|
| `.2f` | 소수점 2자리 실수 | `{3.14159:.2f}` → `3.14` |
| `8.2f` | 너비 8, 소수점 2자리 (오른쪽 정렬) | `{3.14:8.2f}` → `    3.14` |
| `<8.2f` | 왼쪽 정렬 | `{3.14:<8.2f}` → `3.14    ` |
| `d` | 정수 | `{42:d}` → `42` |

```python
# f-string — 기본 삽입 + 서식 지정
name = "김파이썬"
score = 98.7654

# 기본 삽입
print(f"이름: {name}, 점수: {score}")

# 서식 지정
print(f"점수: {score:.2f}")          # 소수점 2자리
print(f"점수: {score:8.2f}")         # 너비 8, 소수점 2자리 (오른쪽 정렬)
print(f"점수: {score:<8.2f}")        # 왼쪽 정렬

# 중괄호 안에 계산식을 직접 넣을 수 있다
a, b = 7, 3
print(f"{a} + {b} = {a + b}")
```

```python
# format() 메서드

# 순서대로 삽입 — {} 에 순서대로 채워진다
print("이름: {}, 나이: {}".format("이파이썬", 30))

# 인덱스로 삽입 — 같은 값을 여러 번 써야 할 때 유용하다
print("{0}이 {1}보다 크다: {2}".format(10, 5, 10 > 5))

# 이름으로 삽입 — 순서에 관계없이 키워드로 지정
print("{city}에 사는 {name}".format(name="박파이썬", city="부산"))
```

```python
# % 포매팅 (구형 방식)
# 기존 코드에서 만날 수 있으니 읽을 수 있어야 하지만, 새 코드에서는 f-string을 쓴다

name = "최파이썬"
score = 95.5

print("이름: %s, 점수: %.1f" % (name, score))
# %s: 문자열, %d: 정수, %f: 실수
# 현재는 f-string 사용을 권장한다
```

## 6. 멀티라인 문자열

- `"""` 또는 `'''` 로 감싸면 줄바꿈을 포함한 여러 줄 문자열을 만들 수 있다
- 주로 두 가지 용도로 쓰인다:
  1. 긴 텍스트 블록을 그대로 변수에 담을 때
  2. 함수나 클래스의 설명을 적는 **독스트링(docstring)** — 함수 본문 첫 줄에 위치하며 `함수명.__doc__` 로 읽을 수 있다

```python
# 여러 줄 문자열
poem = """
파이썬은 읽기 쉽고
배우기 쉬운 언어다.
초보자에게 가장 좋은 선택.
"""
print(poem)

# 독스트링 — 함수 설명을 코드와 함께 관리한다
def add(a, b):
    """
    두 수를 더해 반환한다.

    Args:
        a: 첫 번째 숫자
        b: 두 번째 숫자
    Returns:
        두 수의 합
    """
    return a + b

print(add.__doc__)   # 독스트링을 직접 출력할 수 있다
```
