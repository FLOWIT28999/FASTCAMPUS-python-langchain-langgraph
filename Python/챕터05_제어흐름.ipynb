## 1. 조건문 (if / elif / else)

조건문은 **조건이 참(True)이면 이 코드를, 거짓(False)이면 저 코드를** 실행하도록 분기하는 구문이다.

### 기본 구조

```
if 조건:
    # 조건이 True일 때 실행
elif 다른_조건:
    # 첫 조건은 False, 이 조건이 True일 때 실행
else:
    # 모든 조건이 False일 때 실행
```

> **들여쓰기(indentation)**: 파이썬은 `{}`가 아닌 **4칸 들여쓰기**로 블록을 구분한다.  
> 들여쓰기가 맞지 않으면 `IndentationError`가 발생하므로 주의한다.

### Truthy / Falsy

파이썬에서는 `True`/`False` 외에도 다른 값이 조건식에 들어올 수 있다.  
아래 값들은 **False로 평가**된다.

| 값 | 타입 | 평가 |
|---|---|---|
| `0` | int | False |
| `0.0` | float | False |
| `""` | str | False |
| `[]` | list | False |
| `{}` | dict | False |
| `None` | NoneType | False |

그 외의 값(0이 아닌 숫자, 비어 있지 않은 컬렉션 등)은 모두 **True로 평가**된다.

````python
# if 단독 사용
score = 75

if score >= 60:
    print("합격")

# if-else
temperature = 15
if temperature >= 20:
    print("따뜻합니다")
else:
    print("쌀쌀합니다")
````

````python
# if-elif-else: 여러 조건 중 하나를 선택
score = 82

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"점수: {score}, 학점: {grade}")
````

````python
# 중첩 조건문: if 안에 또 다른 if를 사용
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("입장 가능합니다")
    else:
        print("신분증이 필요합니다")
else:
    print("미성년자는 입장할 수 없습니다")
````

````python
# Truthy / Falsy 확인
# 아래 값들이 조건식에서 어떻게 평가되는지 직접 확인한다
falsy_values = [0, 0.0, "", [], {}, None, False]

for val in falsy_values:
    if val:
        print(f"{repr(val)} → True")
    else:
        print(f"{repr(val)} → False")
````

## 2. while 반복문

`while`은 **조건이 참(True)인 동안 블록을 계속 반복**하는 구문이다.

### 기본 구조

```
while 조건:
    반복할 코드
```

> **무한 루프 주의**: 조건이 절대 `False`가 되지 않으면 프로그램이 영원히 멈추지 않는다.  
> 반복 횟수를 제어하는 **카운터 변수**를 반드시 블록 안에서 증가시켜야 한다.

### 카운터 패턴

```python
count = 0          # 1. 초기화
while count < 5:   # 2. 조건 검사
    print(count)   # 3. 반복 작업
    count += 1     # 4. 카운터 증가 (없으면 무한 루프!)
```

````python
# while 기본: 카운터로 반복 횟수 제어
count = 1
while count <= 5:
    print(f"반복 {count}회")
    count += 1   # 카운터 증가 — 없으면 무한 루프!
````

````python
# while로 1부터 100까지 합계 계산
total = 0
num = 1
while num <= 100:
    total += num
    num += 1

print("1~100 합계:", total)  # 5050
````

## 3. for 반복문과 range()

`for`는 **시퀀스(리스트, 문자열, range 등)의 요소를 하나씩 꺼내 반복**하는 구문이다.  
반복 횟수가 정해져 있을 때 `while`보다 더 간결하게 표현할 수 있다.

### range() 세 가지 형태

| 형태 | 의미 | 예시 | 결과 |
|------|------|------|------|
| `range(stop)` | 0부터 stop-1까지 | `range(5)` | 0, 1, 2, 3, 4 |
| `range(start, stop)` | start부터 stop-1까지 | `range(2, 6)` | 2, 3, 4, 5 |
| `range(start, stop, step)` | step 간격으로 | `range(0, 10, 2)` | 0, 2, 4, 6, 8 |

### enumerate()

리스트를 순회할 때 **인덱스와 값을 동시에** 필요로 하는 경우 `enumerate()`를 사용한다.

```python
for i, value in enumerate(리스트):
    print(i, value)   # i: 인덱스, value: 요소 값
```

````python
# for + range 기본

# 0부터 4까지
for i in range(5):
    print(i, end=" ")
print()  # 줄바꿈

# 1부터 10까지 홀수만 (step=2)
for i in range(1, 11, 2):
    print(i, end=" ")
print()
````

````python
# 리스트 순회
fruits = ["사과", "바나나", "포도"]
for fruit in fruits:
    print(fruit)

print()

# enumerate로 인덱스 + 값 동시에 꺼내기
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

print()

# 문자열 순회: 문자 하나씩 꺼냄
for char in "Python":
    print(char, end="-")
print()

print()

# 딕셔너리 순회: .items()로 키-값 쌍을 동시에 꺼냄
person = {"name": "김파이썬", "age": 25}
for key, value in person.items():
    print(f"{key}: {value}")
````

## 4. 반복 제어 (break / continue / pass)

반복문 안에서 흐름을 조절하는 세 가지 키워드가 있다.

| 키워드 | 동작 | 주로 쓰는 상황 |
|--------|------|----------------|
| `break` | 반복문 즉시 탈출 | 원하는 값을 찾았을 때 |
| `continue` | 현재 반복만 건너뛰고 다음으로 | 특정 조건의 요소는 처리하지 않을 때 |
| `pass` | 아무것도 하지 않음 (자리 채우기) | 나중에 채울 빈 블록이 필요할 때 |

````python
# break: 조건을 만족하는 즉시 반복문 탈출
# 리스트에서 첫 번째 짝수 찾기
numbers = [1, 3, 7, 4, 9, 2]

for n in numbers:
    if n % 2 == 0:
        print("첫 번째 짝수:", n)
        break   # 찾았으면 더 이상 반복할 필요 없음
````

````python
# continue: 현재 반복만 건너뛰고 다음 반복으로 이동
# 1~10 중 홀수만 출력 (짝수는 건너뜀)
for i in range(1, 11):
    if i % 2 == 0:
        continue   # 짝수이면 아래 print를 건너뜀
    print(i, end=" ")
print()
````

````python
# pass: 아무것도 하지 않음 — 문법상 블록이 필요하지만 내용은 나중에 채울 때 사용
for i in range(5):
    if i == 3:
        pass   # 나중에 여기에 특별 처리 추가 예정
    else:
        print(i, end=" ")
print()
````

## 5. 심화: 중첩 반복문 / for·while else

### 중첩 반복문

`for` 안에 또 다른 `for`를 넣을 수 있다.  
외부 반복문이 1번 실행될 때, 내부 반복문은 **처음부터 끝까지 전체를 반복**한다.

```
for 외부_변수 in 외부_시퀀스:
    for 내부_변수 in 내부_시퀀스:
        # 외부 × 내부 횟수만큼 실행
```

### for/else 와 while/else

파이썬의 반복문에는 `else` 블록을 붙일 수 있다.  
**`break` 없이 정상 종료**되었을 때만 `else` 블록이 실행된다.

```python
for 변수 in 시퀀스:
    if 조건:
        break
else:
    # break 없이 반복이 끝났을 때만 실행
```

| 상황 | else 실행 여부 |
|------|----------------|
| 반복이 `break` 없이 끝남 | 실행됨 |
| 반복 중 `break`로 탈출 | 실행 안 됨 |

````python
# 중첩 반복문: 구구단 (2~4단)
for dan in range(2, 5):        # 외부: 2단, 3단, 4단
    for num in range(1, 10):   # 내부: 1~9
        print(f"{dan} x {num} = {dan * num}")
    print()   # 단 사이 빈 줄
````

````python
# for/else: break로 탈출하면 else 실행 안 됨
target = "포도"
fruits = ["사과", "바나나", "딸기"]

for fruit in fruits:
    if fruit == target:
        print(f"{target} 발견!")   # break 발생 → else 건너뜀
        break
else:
    # break 없이 반복이 끝났을 때만 실행
    print(f"{target}이(가) 목록에 없습니다")

print()

# 반대로: target이 목록에 있는 경우
target2 = "바나나"
for fruit in fruits:
    if fruit == target2:
        print(f"{target2} 발견!")   # break 실행 → else 블록 건너뜀
        break
else:
    print(f"{target2}이(가) 목록에 없습니다")
````

## 챕터 정리

### 이번 챕터에서 배운 내용

- `if / elif / else`로 조건에 따른 코드 분기를 작성한다.
- 파이썬의 **들여쓰기(4칸)**가 블록을 구분하는 핵심 문법이다.
- `0`, `""`, `[]`, `None` 등은 **Falsy** 값으로 조건식에서 `False`로 평가된다.
- `while`은 조건이 참인 동안 반복하고, `for`는 시퀀스 요소를 순서대로 처리한다.
- `range(start, stop, step)`으로 숫자 범위를 간편하게 생성한다.
- `enumerate()`로 인덱스와 값을 동시에 순회한다.
- `break`(즉시 탈출), `continue`(현재 반복 건너뜀), `pass`(빈 자리 채우기)로 반복 흐름을 제어한다.
- `for/else`, `while/else`에서 `else`는 `break` 없이 정상 종료 시에만 실행된다.

### 핵심 키워드 한눈에 보기

| 분류 | 키워드/함수 | 역할 |
|------|-------------|------|
| 조건문 | `if`, `elif`, `else` | 조건에 따라 실행 경로 선택 |
| while 반복 | `while` | 조건이 True인 동안 반복 |
| for 반복 | `for`, `in` | 시퀀스 요소를 순서대로 처리 |
| 범위 생성 | `range()` | 정수 범위를 만들어 for에 활용 |
| 인덱스 + 값 | `enumerate()` | 인덱스와 값을 동시에 순회 |
| 반복 제어 | `break` | 반복문 즉시 탈출 |
| 반복 제어 | `continue` | 현재 반복 건너뛰고 다음으로 |
| 반복 제어 | `pass` | 빈 블록 자리 채우기 |
| 정상 종료 감지 | `for/while ... else` | break 없이 끝났을 때 실행 |
