## 1. 변수란 무엇인가

변수(Variable)는 **데이터를 담는 이름표**입니다.

예를 들어, 편의점에서 물건에 가격표를 붙이는 것처럼, 우리도 데이터에 이름을 붙여서 나중에 다시 꺼내 쓸 수 있습니다.

```
name = "홍길동"
  |        |
 이름표   저장할 값
```

### `=` 연산자의 의미

`=`는 수학의 "같다"가 아닙니다. Python에서 `=`는 **"오른쪽 값을 왼쪽 이름에 저장하라"** 는 명령입니다.

```
age = 25
     ↑
   오른쪽 값(25)을 왼쪽 이름(age)에 저장
```

### 변수를 사용하는 이유

- 값을 기억해두고 나중에 다시 꺼내 쓸 수 있습니다
- 같은 값을 여러 곳에서 사용할 때, 변수 하나만 바꾸면 전체가 반영됩니다
- 코드에 의미 있는 이름을 붙여 읽기 쉽게 만들 수 있습니다

````python
# 변수에 값을 저장하고 출력하기
name = "홍길동"
age = 27
print(name)
print(age)
````

````python
# 변수는 나중에 값을 바꿀 수 있습니다
score = 80
print("처음 점수:", score)

score = 95
print("바뀐 점수:", score)
````

## 2. 변수 명명 규칙

변수 이름을 지을 때는 반드시 지켜야 하는 **규칙**과, 지키면 좋은 **관례**가 있습니다.

### 규칙 (반드시 지켜야 함)

| 규칙 | 올바른 예 | 잘못된 예 |
|------|-----------|----------|
| 영문자, 숫자, 밑줄(`_`)만 사용 | `user_name` | `user-name` |
| 숫자로 시작 불가 | `score1` | `1score` |
| 공백 불가 | `total_score` | `total score` |
| 대소문자 구분 | `Name`과 `name`은 다른 변수 | |
| Python 예약어 사용 불가 | `if_result` | `if`, `for`, `while` |

### 관례 (권장)

- **snake_case** 사용: 단어 사이를 밑줄로 연결합니다 → `user_name`, `total_score`
- **의미 있는 이름** 사용: `x`보다는 `width`, `a`보다는 `age`가 훨씬 읽기 좋습니다

> Python 예약어란 if, for, while처럼 Python이 이미 특별한 용도로 사용하는 단어입니다. 변수 이름으로 사용하면 오류가 발생합니다.

````python
# 좋은 변수 이름 예시
user_name = "김파이썬"
total_score = 100
is_student = True

print(user_name, total_score, is_student)
````

## 3. 숫자형 (int, float, complex)

Python에는 세 가지 숫자 자료형이 있습니다.

| 자료형 | 이름 | 설명 | 예시 |
|--------|------|------|------|
| `int` | 정수 | 소수점이 없는 숫자 (양수, 음수, 0) | `5`, `-3`, `0` |
| `float` | 실수 | 소수점이 있는 숫자 | `3.14`, `175.5` |
| `complex` | 복소수 | 실수부 + 허수부 | `3+4j` |

### int와 float를 구분하는 이유

- **정확한 정수 계산**이 필요할 때는 `int`를 사용합니다 (개수 세기, 인덱스 등)
- **소수점이 필요한 계산**에는 `float`를 사용합니다 (평균, 비율, 측정값 등)
- Python에서 `/` 나누기는 항상 `float`를 반환합니다 (10 / 2 = 5.0)
- `//` 연산자를 사용하면 몫만 구해 `int`로 반환합니다 (10 // 3 = 3)

````python
# 정수 (int)
apples = 5
temperature = -3
zero = 0
print(apples, temperature, zero)
````

````python
# 실수 (float)
height = 175.5
pi = 3.14159
print(height, pi)

# 정수와 실수의 나누기
print(10 / 3)   # 나누기는 항상 float 반환
print(10 // 3)  # 몫만 구하면 int 반환
````

````python
# 복소수 (complex)
c = 3 + 4j
print(c)
print("실수부:", c.real)
print("허수부:", c.imag)
````

## 4. 문자열 (str)

문자열(String)은 **따옴표로 감싼 텍스트 데이터**입니다.

- 큰따옴표(`"`) 또는 작은따옴표(`'`) 모두 사용할 수 있습니다
- 따옴표로 감싸면 숫자처럼 보여도 문자열입니다: `"123"` ≠ `123`
- `+` 연산자로 문자열을 이어붙일 수 있습니다 (연결)
- `*` 연산자로 문자열을 반복할 수 있습니다

````python
greeting = "안녕하세요"
name = '파이썬'

print(greeting)
print(name)

# 문자열 연결
print(greeting + ", " + name + "!")
````

````python
# 문자열 반복
line = "-" * 20
print(line)
print("Python")
print(line)
````

````python
# 숫자 문자열과 정수는 다릅니다
num_str = "123"
num_int = 123

print(num_str + num_str)  # 문자열 연결 → "123123"
print(num_int + num_int)  # 숫자 덧셈  → 246
````

## 5. 불리언 (bool)

불리언(Boolean)은 **`True`(참) 또는 `False`(거짓)** 두 가지 값만 가지는 자료형입니다.

- 반드시 대문자로 시작해야 합니다: `True`, `False` (소문자 `true`는 오류)
- 비교 연산(`>`, `<`, `==`, `!=`)의 결과는 항상 불리언입니다
- 조건에 따라 다르게 동작하는 코드를 작성할 때 활용합니다 (챕터 05에서 본격적으로 다룹니다)

| 비교 연산자 | 의미 |
|-------------|------|
| `==` | 같다 |
| `!=` | 다르다 |
| `>` | 크다 |
| `<` | 작다 |
| `>=` | 크거나 같다 |
| `<=` | 작거나 같다 |

````python
is_raining = True
is_sunny = False

print(is_raining)
print(is_sunny)

# 비교 연산의 결과는 bool
print(10 > 5)    # True
print(10 < 5)    # False
print(10 == 10)  # True
````

## 6. None

`None`은 **"아무것도 없음"을 나타내는 특별한 값**입니다.

- 빈 문자열(`""`)이나 숫자 `0`과는 다릅니다
  - `""`: 빈 문자열 — 그릇은 있는데 내용이 없는 상태
  - `0`: 숫자 0 — 값이 0이라는 의미
  - `None`: 그릇 자체가 없는 상태, 아예 정의되지 않음
- 변수를 "비어있음"으로 초기화할 때 사용합니다
- 함수가 명시적으로 값을 반환하지 않을 때 자동으로 `None`이 반환됩니다
- `None`인지 확인할 때는 `==` 대신 `is`를 사용하는 것이 관례입니다

````python
result = None
print(result)
print(result is None)  # None인지 확인할 때는 == 대신 is를 사용
````

## 7. type() 함수로 자료형 확인하기

`type(값)` 함수는 해당 값이 **어떤 자료형인지 반환**합니다.

- 처음 보는 변수의 자료형을 확인할 때 유용합니다
- 자료형에 따라 사용할 수 있는 연산이 달라지므로, 오류 디버깅에도 자주 활용됩니다

````python
print(type(42))       # int
print(type(3.14))     # float
print(type(2 + 3j))   # complex
print(type("안녕"))    # str
print(type(True))     # bool
print(type(None))     # NoneType
````

````python
# 변수에 담아서 확인하기
data = 100
print(data, "의 자료형:", type(data))

data = "백"
print(data, "의 자료형:", type(data))
````

````python
# 자료형끼리 섞으면 오류가 납니다
# number = 10 + "20"  # 이 줄의 주석을 해제하면 TypeError 발생
# print(number)

# 형변환(Type Casting)으로 해결할 수 있습니다
number = str(10) + int("20")   # 문자열 "20"을 정수 20으로 변환
print(number)

text = "현재 점수: " + str(95)  # 정수 95를 문자열 "95"로 변환
print(text)
````

## 챕터 정리

이번 챕터에서 배운 내용을 정리합니다.

- **변수**: 데이터를 담는 이름표입니다. `=` 연산자로 오른쪽 값을 왼쪽 이름에 저장하며, 나중에 값을 바꿀 수 있습니다
- **명명 규칙**: 영문자/숫자/밑줄만 사용하고, 숫자로 시작하지 않으며, 예약어는 피합니다. 관례로 `snake_case`와 의미 있는 이름을 사용합니다
- **숫자형**: `int`(정수), `float`(실수), `complex`(복소수)가 있으며, `/`는 항상 float, `//`는 정수 몫을 반환합니다
- **문자열(`str`)**: 따옴표로 감싼 텍스트입니다. `+`로 연결하고, `*`로 반복합니다
- **불리언(`bool`)**: `True` / `False` 두 가지 값이며, 비교 연산의 결과로 나타납니다
- **None**: 값이 아예 없음을 나타냅니다. `is None`으로 확인합니다
- **type()**: 변수나 값의 자료형을 확인합니다. 형변환 함수(`int()`, `float()`, `str()`)로 자료형을 바꿀 수 있습니다

### 자료형 한눈에 비교

| 자료형 | 예시 | 특징 |
|--------|------|------|
| `int` | `42`, `-3`, `0` | 소수점 없는 정수 |
| `float` | `3.14`, `175.5` | 소수점 있는 실수 |
| `complex` | `3+4j` | 실수부 + 허수부 |
| `str` | `"안녕"`, `'Python'` | 따옴표로 감싼 텍스트 |
| `bool` | `True`, `False` | 참/거짓 두 가지 값 |
| `NoneType` | `None` | 값이 없음을 나타냄 |
