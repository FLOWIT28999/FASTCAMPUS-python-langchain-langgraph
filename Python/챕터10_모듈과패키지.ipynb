## 1. 모듈과 패키지 개념

### 모듈(Module)

`.py` 파일 하나가 곧 **모듈**이다. 변수, 함수, 클래스를 묶어 놓은 파일로,
다른 파이썬 파일에서 `import`해 재사용할 수 있다.

```
my_utils.py       <-- 이 파일 자체가 모듈
```

### 패키지(Package)

여러 모듈을 담은 **디렉터리**가 패키지다.  
디렉터리 안에 `__init__.py` 파일이 있어야 파이썬이 이 폴더를 패키지로 인식한다.

```
my_package/
    __init__.py      <-- 패키지 인식 파일 (비어 있어도 됨)
    utils.py
    network.py
```

### import 3가지 방법

| 방식 | 예시 | 특징 |
|------|------|------|
| `import 모듈명` | `import math` | `모듈명.함수명()`으로 접근, 출처가 명확 |
| `from 모듈명 import 이름` | `from math import sqrt` | 모듈명 없이 바로 사용, 자주 쓰는 함수에 편리 |
| `import 모듈명 as 별칭` | `import random as rd` | 긴 모듈명을 짧게 줄임 |

````python
# 방식 1: import 모듈명 — 모듈명.함수명() 으로 접근
import math
print(f"math.sqrt(16) = {math.sqrt(16)}")
print(f"math.pi = {math.pi:.6f}")
````

````python

print()

# 방식 2: from 모듈명 import 이름 — 모듈명 없이 바로 사용
from math import ceil, floor
print(f"ceil(3.2)  = {ceil(3.2)}")
print(f"floor(3.9) = {floor(3.9)}")

print()

# 방식 3: import 모듈명 as 별칭 — 짧은 이름으로 사용
import random as rd
rd.seed(42)   # 결과를 고정해 재현성 확보
print(f"rd.randint(1, 10) = {rd.randint(1, 10)}")
````

## 2. 표준 라이브러리 — math / random

파이썬을 설치하면 **별도 설치 없이** 바로 쓸 수 있는 라이브러리다.

### math 주요 함수와 상수

| 함수 / 상수 | 설명 | 예시 |
|-------------|------|------|
| `math.sqrt(x)` | 제곱근 | `sqrt(9)` → `3.0` |
| `math.ceil(x)` | 올림 | `ceil(3.1)` → `4` |
| `math.floor(x)` | 내림 | `floor(3.9)` → `3` |
| `math.pow(x, y)` | 거듭제곱 | `pow(2, 10)` → `1024.0` |
| `math.log(x)` | 자연로그 | `log(math.e)` → `1.0` |
| `math.factorial(n)` | 팩토리얼 | `factorial(5)` → `120` |
| `math.pi` | 원주율 π | `3.14159...` |
| `math.e` | 자연상수 e | `2.71828...` |

### random 주요 함수

| 함수 | 설명 |
|------|------|
| `random.random()` | 0.0 이상 1.0 미만 실수 |
| `random.randint(a, b)` | a 이상 b 이하 정수 |
| `random.choice(seq)` | 시퀀스에서 임의 요소 1개 |
| `random.shuffle(seq)` | 시퀀스를 제자리에서 섞기 |
| `random.sample(seq, k)` | 중복 없이 k개 추출 (원본 유지) |

````python
import math

print("=== math 모듈 ===")
print(f"sqrt(144)      = {math.sqrt(144)}")
print(f"ceil(7.1)      = {math.ceil(7.1)}")
print(f"floor(7.9)     = {math.floor(7.9)}")
print(f"pow(2, 10)     = {math.pow(2, 10):.0f}")
print(f"log(math.e)    = {math.log(math.e)}")
print(f"factorial(10)  = {math.factorial(10)}")
print(f"pi             = {math.pi:.10f}")
print(f"e              = {math.e:.10f}")

print()

# 활용 예: 원의 넓이 계산
반지름 = 5
넓이 = math.pi * math.pow(반지름, 2)
print(f"반지름 {반지름}인 원의 넓이: {넓이:.4f}")
````

````python
import random

random.seed(7)   # 실행마다 동일한 결과를 얻기 위해 시드 고정

print("=== random 모듈 ===")
print(f"random()           = {random.random():.4f}")
print(f"randint(1, 100)    = {random.randint(1, 100)}")

과일 = ["사과", "바나나", "포도", "딸기", "수박"]
print(f"choice(과일)       = {random.choice(과일)}")

카드 = list(range(1, 14))
random.shuffle(카드)
print(f"shuffle 결과       = {카드}")

로또번호 = sorted(random.sample(range(1, 46), 6))
print(f"로또 번호          = {로또번호}")
````

## 3. 표준 라이브러리 — datetime

`datetime` 모듈은 날짜와 시간을 다루는 클래스들을 제공한다.

| 클래스 | 설명 |
|--------|------|
| `date` | 날짜만 (년, 월, 일) |
| `datetime` | 날짜 + 시간 |
| `timedelta` | 시간 간격 (두 날짜의 차이) |

- `strftime()`: datetime → 문자열 포매팅
- `strptime()`: 문자열 → datetime 변환

### 주요 포맷 코드

| 포맷 코드 | 의미 | 예시 |
|-----------|------|------|
| `%Y` | 4자리 연도 | `2026` |
| `%m` | 월 (01~12) | `02` |
| `%d` | 일 (01~31) | `26` |
| `%H` | 시 (00~23) | `14` |
| `%M` | 분 (00~59) | `30` |

````python
from datetime import datetime, date, timedelta

# 오늘 날짜와 현재 시각
오늘 = date.today()
지금 = datetime.now()
print(f"오늘 날짜: {오늘}")
print(f"현재 시각: {지금.strftime('%Y-%m-%d %H:%M:%S')}")

print()

# timedelta — 날짜 계산
한달뒤 = 오늘 + timedelta(days=30)
일주일전 = 오늘 - timedelta(weeks=1)
print(f"30일 후: {한달뒤}")
print(f"1주일 전: {일주일전}")

print()

# 날짜 차이 계산 (D-Day)
목표일 = date(2026, 12, 31)
남은일 = (목표일 - 오늘).days
print(f"2026년 12월 31일까지: D-{남은일}")

print()

# 문자열 → datetime 변환
날짜_문자열 = "2026-03-01"
변환된날짜 = datetime.strptime(날짜_문자열, "%Y-%m-%d").date()
print(f"'{날짜_문자열}' 파싱 결과: {변환된날짜} (타입: {type(변환된날짜).__name__})")
````

## 4. 표준 라이브러리 — os / sys

### os 모듈

운영체제와 상호작용하는 기능을 제공한다 — 파일/디렉터리 관리, 경로 처리 등.

| 함수 | 설명 |
|------|------|
| `os.getcwd()` | 현재 작업 디렉터리 경로 반환 |
| `os.listdir(경로)` | 디렉터리 내 파일 목록 |
| `os.path.join(a, b)` | 경로 문자열 안전하게 결합 |
| `os.path.exists(경로)` | 경로 존재 여부 확인 |
| `os.path.isfile(경로)` | 파일인지 확인 |
| `os.makedirs(경로)` | 디렉터리 생성 (중간 경로 포함) |

### sys 모듈

파이썬 인터프리터 자체의 정보와 동작을 제어한다.

| 속성 / 함수 | 설명 |
|-------------|------|
| `sys.version` | 파이썬 버전 문자열 |
| `sys.platform` | 운영체제 식별자 (`win32`, `darwin`, `linux`) |
| `sys.path` | 모듈 탐색 경로 목록 |
| `sys.modules` | 현재 import된 모듈 딕셔너리 |
| `sys.exit()` | 프로그램 종료 |

````python
import os

print("=== os 모듈 ===")
현재경로 = os.getcwd()
print(f"현재 작업 디렉터리: {현재경로}")

print()

# 파일 목록 (.ipynb 파일만 필터링)
파일목록 = os.listdir(".")
노트북목록 = [f for f in 파일목록 if f.endswith(".ipynb")]
print(f"현재 폴더의 .ipynb 파일 ({len(노트북목록)}개):")
for 파일 in sorted(노트북목록):
    print(f"  {파일}")

print()

# 경로 결합 — os.path.join은 운영체제에 맞는 구분자를 자동 사용
경로 = os.path.join(현재경로, "sample.txt")
print(f"경로 결합 결과: {경로}")
print(f"sample.txt 존재 여부: {os.path.exists('sample.txt')}")
````

````python
import sys

print("=== sys 모듈 ===")
print(f"Python 버전: {sys.version.split()[0]}")
print(f"플랫폼: {sys.platform}")

print()

# sys.path — 파이썬이 모듈을 찾는 경로 목록
print("모듈 탐색 경로 (sys.path) — 처음 3개:")
for 경로 in sys.path[:3]:
    print(f"  {경로}")

print()

# 현재 import된 모듈 수 확인
print(f"현재 import된 모듈 수: {len(sys.modules)}개")
````

## 5. pip과 외부 패키지

`pip`은 파이썬의 공식 패키지 관리자다.  
[PyPI(Python Package Index)](https://pypi.org/)에 등록된 수십만 개의 패키지를 설치·삭제할 수 있다.

### 주요 pip 명령어

| 명령어 | 설명 |
|--------|------|
| `pip install 패키지명` | 패키지 설치 |
| `pip install 패키지명==버전` | 특정 버전 설치 |
| `pip uninstall 패키지명` | 패키지 삭제 |
| `pip list` | 설치된 패키지 목록 |
| `pip show 패키지명` | 패키지 상세 정보 |
| `pip install -r requirements.txt` | 목록 파일에서 일괄 설치 |

### requirements.txt

프로젝트에 필요한 패키지 목록을 기록한 텍스트 파일이다.  
팀원과 동일한 환경을 공유하거나 서버에 배포할 때 사용한다.

```
requirements.txt 예시
----------------------
requests==2.31.0
numpy==1.26.0
pandas==2.1.0
```

Jupyter 노트북에서는 셀 맨 앞에 `!`를 붙이면 터미널 명령어처럼 실행할 수 있다.

```python
!pip install requests
!pip install -r requirements.txt
```

````python
import sys
import subprocess

# pip 버전 확인
결과 = subprocess.run(
    [sys.executable, "-m", "pip", "--version"],
    capture_output=True, text=True
)
print(결과.stdout.strip())

print()

# 설치된 패키지 목록 (처음 10개)
결과2 = subprocess.run(
    [sys.executable, "-m", "pip", "list"],
    capture_output=True, text=True
)
라인들 = 결과2.stdout.strip().split("\n")
print(f"설치된 패키지 수: {len(라인들) - 2}개")
print("\n처음 10개 패키지:")
for 라인 in 라인들[2:12]:   # 헤더 2줄 건너뜀
    print(f"  {라인}")

print()

# requirements.txt 파일 예시 생성
requirements = "requests==2.31.0\nnumpy==1.26.0\npandas==2.1.0\n"
with open("requirements_example.txt", "w") as f:
    f.write(requirements)
print("requirements_example.txt 생성 완료")
print("내용:")
print(requirements)
print("설치 명령어: pip install -r requirements_example.txt")
````

## 6. 직접 모듈 만들기

파이썬 파일(`.py`)을 직접 작성하면 곧바로 **모듈**이 된다.  
함수와 변수를 모듈로 분리하면 여러 프로젝트에서 재사용할 수 있다.

### 모듈 파일 작성 단계

```
1. my_utils.py 파일 작성 (함수, 변수 정의)
2. 같은 디렉터리의 다른 .py 또는 .ipynb에서 import my_utils
3. my_utils.함수명() 형태로 사용
```

### `__init__.py` 역할

여러 모듈을 하나의 **패키지**로 묶으려면 디렉터리 안에 `__init__.py` 파일이 있어야 한다.

- 비어 있어도 되고, 패키지 초기화 코드를 넣어도 된다.
- 패키지를 import할 때 `__init__.py`가 먼저 실행된다.
- 파이썬 3.3 이후로는 없어도 일부 상황에서 동작하지만, 명시적으로 작성하는 것이 관례다.

> 아래 Code 8을 먼저 실행해 `my_utils.py`를 생성한 뒤, Code 9에서 import 하세요.

````python
# 파이썬 파일(.py)을 직접 만들면 곧바로 모듈이 된다
모듈_코드 = '''# my_utils.py — 직접 만든 유틸리티 모듈

def 인사(이름):
    """이름을 받아 인사 메시지를 반환한다."""
    return f"안녕하세요, {이름}님!"

def 평균(숫자목록):
    """숫자 리스트의 평균을 반환한다."""
    if not 숫자목록:
        return 0
    return sum(숫자목록) / len(숫자목록)

def 최댓값_위치(숫자목록):
    """리스트에서 최댓값의 인덱스를 반환한다."""
    return 숫자목록.index(max(숫자목록))

버전 = "1.0.0"
작성자 = "파이썬 학습자"
'''

with open("my_utils.py", "w", encoding="utf-8") as f:
    f.write(모듈_코드)

print("my_utils.py 파일 생성 완료")
print()
with open("my_utils.py", "r", encoding="utf-8") as f:
    print(f.read())
````

````python
# Code 8을 먼저 실행해 my_utils.py를 생성해야 합니다
import my_utils

# 모듈명.함수명() 방식
print(my_utils.인사("홍길동"))
점수들 = [85, 92, 78, 96, 88]
print(f"평균 점수: {my_utils.평균(점수들):.1f}")
print(f"최고 점수 위치: {my_utils.최댓값_위치(점수들)}번 인덱스")
print(f"버전: {my_utils.버전}")
print(f"작성자: {my_utils.작성자}")

print()

# from ... import 방식
from my_utils import 인사, 평균
print(인사("이순신"))
print(f"평균: {평균([100, 90, 85, 92]):.1f}")

print()

# 모듈의 공개 목록 확인
공개목록 = [이름 for 이름 in dir(my_utils) if not 이름.startswith("_")]
print(f"my_utils 공개 목록: {공개목록}")
````

````python
# 종합 실습 — random + datetime 활용한 D-Day 계산기
import random
from datetime import datetime, date, timedelta


def 행운의_번호(개수=5, 최대=45):
    """중복 없는 행운의 번호를 뽑는다."""
    return sorted(random.sample(range(1, 최대 + 1), 개수))


def 디데이(목표_문자열):
    """'YYYY-MM-DD' 형식의 날짜까지 남은 일수를 반환한다."""
    try:
        목표 = datetime.strptime(목표_문자열, "%Y-%m-%d").date()
    except ValueError:
        raise ValueError(f"날짜 형식이 잘못되었습니다: {목표_문자열} (올바른 형식: YYYY-MM-DD)")
    return (목표 - date.today()).days


def 디데이_메시지(이름, 날짜_문자열):
    """D-Day 정보를 문자열로 반환한다."""
    남은 = 디데이(날짜_문자열)
    if 남은 > 0:
        return f"D-{남은:4d}  {이름} ({날짜_문자열})"
    elif 남은 == 0:
        return f"D-Day  {이름} ({날짜_문자열}) — 오늘입니다!"
    else:
        return f"D+{abs(남은):4d}  {이름} ({날짜_문자열}) — {abs(남은)}일 경과"


# 실행
요일_이름 = ["월", "화", "수", "목", "금", "토", "일"]
오늘 = date.today()
print(f"오늘: {오늘.strftime('%Y년 %m월 %d일')} ({요일_이름[오늘.weekday()]}요일)")
print(f"행운의 번호: {행운의_번호()}")
print()

# D-Day 계산
이벤트들 = [
    ("연말",         "2026-12-31"),
    ("여름 방학",    "2026-07-20"),
    ("챕터 12 완성", "2026-03-15"),
]

print("=== D-Day 현황 ===")
for 이름, 날짜 in 이벤트들:
    try:
        print(디데이_메시지(이름, 날짜))
    except ValueError as e:
        print(f"[오류] {e}")
````

## 챕터 10 정리

### 이번 챕터에서 배운 내용

- **모듈**: `.py` 파일 하나, 변수·함수·클래스의 묶음
- **패키지**: 모듈을 담은 디렉터리 + `__init__.py`
- **import 3가지 방식**: `import`, `from ... import`, `import ... as`
- **표준 라이브러리**: 설치 없이 바로 쓸 수 있는 내장 모듈
- **pip**: 외부 패키지 설치·관리 도구, `requirements.txt`로 환경 공유
- **커스텀 모듈**: `.py` 파일 작성 → 같은 디렉터리에서 import

### import 방식 비교

| 방식 | 예시 | 언제 쓰면 좋은가 |
|------|------|------------------|
| `import 모듈명` | `import math` | 모듈 전체를 쓸 때, 출처를 명확히 할 때 |
| `from 모듈명 import 이름` | `from math import sqrt` | 특정 함수만 자주 쓸 때 |
| `import 모듈명 as 별칭` | `import numpy as np` | 모듈명이 길거나 관례적인 별칭이 있을 때 |

### 표준 라이브러리 한눈에 보기

| 모듈 | 주요 용도 | 대표 기능 |
|------|-----------|----------|
| `math` | 수학 계산 | `sqrt`, `ceil`, `floor`, `pi`, `factorial` |
| `random` | 난수 생성 | `randint`, `choice`, `shuffle`, `sample` |
| `datetime` | 날짜·시간 처리 | `date.today()`, `timedelta`, `strftime` |
| `os` | 파일·경로 관리 | `getcwd`, `listdir`, `path.join`, `makedirs` |
| `sys` | 인터프리터 제어 | `version`, `platform`, `path`, `modules` |
