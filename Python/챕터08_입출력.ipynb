## 1. print() 심화

`print()` 함수는 단순히 값을 화면에 출력하는 것 이상의 기능을 제공합니다.
세 가지 키워드 인자를 조합하면 출력 형식을 세밀하게 제어할 수 있습니다.

| 인자 | 기본값 | 역할 |
|------|--------|------|
| `sep` | `" "` (공백 한 칸) | 여러 값 사이에 삽입되는 구분자 |
| `end` | `"\n"` (줄바꿈) | 마지막 값 뒤에 붙는 문자 |
| `file` | `sys.stdout` (화면) | 출력 대상 파일 객체 |

### sep 인자

```python
print(값1, 값2, 값3, sep="구분자")
```

콤마로 나열한 여러 값 **사이**에 `sep`에 지정한 문자열이 삽입됩니다.
기본값은 공백 한 칸이며 빈 문자열 `""` 로 지정하면 구분자 없이 붙여서 출력합니다.

### end 인자

```python
print("텍스트", end="")
```

기본값인 `"\n"` 대신 다른 문자를 지정하면 출력 후 줄바꿈을 하지 않습니다.
같은 줄에 이어서 출력하거나 진행 표시(`.`, `*`)를 만들 때 활용합니다.

### file 인자

```python
print("텍스트", file=파일객체)
```

열려 있는 파일 객체를 지정하면 화면 대신 파일에 직접 출력합니다.
로그를 파일에 기록하거나 `write()` 없이 편리하게 파일에 텍스트를 쓸 때 사용합니다.

````python
# sep: 여러 값 사이에 들어갈 구분자
print("사과", "바나나", "포도")               # 기본: 공백
print("사과", "바나나", "포도", sep=", ")     # 쉼표+공백
print("2026", "02", "26", sep="-")            # 날짜 형식
print("A", "B", "C", sep="")                  # 구분자 없음
````

````python

print()  # 빈 줄

# end: 줄 끝 문자 변경 (기본값은 \n)
print("로딩 중", end="")
print(".", end="")
print(".", end="")
print(".")           # 마지막은 줄바꿈 (end 기본값 사용)

print()  # 빈 줄

# sep과 end 동시에 사용
for i in range(1, 6):
    print(i, end=" | ")   # 1 | 2 | 3 | 4 | 5 |
print()   # 마지막 줄바꿈
````

````python
# file 인자: 출력 대상을 파일 객체로 지정
with open("print_output.txt", "w", encoding="utf-8") as f:
    print("파일에 직접 출력된 첫 번째 줄", file=f)
    print("두 번째 줄", file=f)
    print("세 항목", "A", "B", sep="|", file=f)   # sep도 함께 적용

# 결과 확인: 방금 만든 파일을 읽어서 화면에 출력
with open("print_output.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

## 2. input()과 형변환

### input() 함수

```python
변수 = input("안내 문구: ")
```

- 화면에 안내 문구를 출력한 뒤 사용자가 Enter 키를 누를 때까지 기다립니다
- 사용자가 입력한 내용을 **문자열(str)** 로 반환합니다
- 반환값은 항상 `str` 타입이므로 숫자로 사용하려면 명시적 형변환이 필요합니다

### 형변환

| 함수 | 변환 대상 | 예시 |
|------|-----------|------|
| `int(값)` | 정수 | `int("42")` → `42` |
| `float(값)` | 실수 | `float("3.14")` → `3.14` |
| `str(값)` | 문자열 | `str(100)` → `"100"` |

### 주의사항

- `int("안녕")` 처럼 숫자로 변환할 수 없는 문자열을 변환하면 **`ValueError`** 가 발생합니다
- 잘못된 입력에 안전하게 대응하는 방법은 **챕터 09 예외 처리**에서 배웁니다

````python
# 이 셀을 실행하면 아래에 입력창이 나타납니다
name = input("이름을 입력하세요: ")
age_str = input("나이를 입력하세요: ")

age = int(age_str)   # 문자열 → 정수
print(f"안녕하세요, {name}님! 내년이면 {age + 1}살이 됩니다.")

print()  # 빈 줄

height = float(input("키(cm)를 입력하세요: "))
print(f"키: {height:.1f}cm")

print()  # 빈 줄

# input()의 반환값은 항상 str임을 확인
num_str = input("숫자를 하나 입력하세요: ")
print(f"입력값의 타입: {type(num_str)}")
num = int(num_str)
print(f"변환 후 타입: {type(num)}, 값 * 2 = {num * 2}")
````

## 3. 파일 입출력 개요

### open() 함수

```python
파일객체 = open(파일경로, 모드, encoding="utf-8")
```

파일을 열어 파일 객체를 반환합니다. 작업이 끝나면 반드시 `close()`를 호출해야 합니다.

### 파일 모드

| 모드 | 의미 | 파일이 없을 때 | 기존 내용 |
|------|------|---------------|-----------|
| `'r'` | 읽기 (기본값) | `FileNotFoundError` | 보존 |
| `'w'` | 쓰기 | 새 파일 생성 | **삭제** |
| `'a'` | 추가 쓰기 | 새 파일 생성 | 보존 (뒤에 이어 씀) |
| `'rb'` / `'wb'` | 바이너리 읽기/쓰기 | (이미지, 음악 등 이진 파일에 사용) | — |

### with 문 — 안전한 파일 다루기

```python
with open("파일명", "r", encoding="utf-8") as f:
    내용 = f.read()
# 블록을 벗어나면 f.close()가 자동 호출됩니다
```

`with` 문을 사용하면 예외가 발생하더라도 파일이 반드시 닫힙니다.
직접 `close()`를 호출하지 않아도 되므로 파일 손상이나 리소스 누수를 방지할 수 있습니다.

### 인코딩

한글 텍스트 파일을 읽고 쓸 때는 `encoding="utf-8"` 을 명시적으로 지정하는 것을 권장합니다.
생략하면 운영체제의 기본 인코딩이 사용되어 플랫폼에 따라 결과가 달라질 수 있습니다.

````python
# 읽기 실습에 사용할 sample.txt 생성 ('w' 모드: 기존 내용 삭제 후 새로 씀)
with open("sample.txt", "w", encoding="utf-8") as f:
    f.write("첫 번째 줄\n")
    f.write("두 번째 줄\n")
    f.write("세 번째 줄\n")

print("sample.txt 파일이 생성되었습니다.")
````

````python

# 파일이 실제로 만들어졌는지 확인
import os
print(f"파일 크기: {os.path.getsize('sample.txt')} 바이트")
````

````python
# read() — 파일 전체를 하나의 문자열로 반환
with open("sample.txt", "r", encoding="utf-8") as f:
    content = f.read()
print("=== read() ===")
print(content)
````

````python


# readlines() — 각 줄을 요소로 갖는 리스트 반환 (줄바꿈 문자 포함)
with open("sample.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
print("=== readlines() ===")
print(lines)
print()
````

````python

# for문으로 한 줄씩 순회 — 대용량 파일에 메모리 효율적
with open("sample.txt", "r", encoding="utf-8") as f:
    print("=== 줄별 순회 ===")
    for line in f:
        print(line.rstrip())   # rstrip()으로 줄 끝 줄바꿈 문자 제거
````

````python
# readline() — 호출할 때마다 다음 줄 하나를 반환
# 파일 끝에 도달하면 빈 문자열 '' 을 반환합니다
with open("sample.txt", "r", encoding="utf-8") as f:
    line1 = f.readline()
    line2 = f.readline()
    line3 = f.readline()
    line4 = f.readline()   # 줄이 없으므로 빈 문자열 반환

# repr()을 사용하면 줄바꿈 문자(\n)와 빈 문자열을 눈으로 확인할 수 있습니다
print(repr(line1))   # '첫 번째 줄\n'
print(repr(line2))   # '두 번째 줄\n'
print(repr(line3))   # '세 번째 줄\n'
print(repr(line4))   # '' — 파일 끝
````

## 4. 파일 쓰기 심화

### 'w' 모드 vs 'a' 모드

| 구분 | `'w'` 모드 | `'a'` 모드 |
|------|-----------|------------|
| 기존 파일이 있을 때 | 내용을 **전부 삭제**하고 처음부터 씀 | 기존 내용 **보존**, 파일 끝에 이어서 씀 |
| 파일이 없을 때 | 새 파일 생성 | 새 파일 생성 |
| 주요 용도 | 덮어쓰기, 새 파일 생성 | 로그 파일 추가, 데이터 누적 |

### write() vs writelines()

```python
f.write("문자열")            # 문자열 하나를 파일에 씁니다
f.writelines(["줄1\n", "줄2\n"])  # 리스트의 각 요소를 이어서 씁니다
```

두 함수 모두 줄바꿈을 자동으로 삽입하지 않습니다.
줄 단위로 구분하려면 각 문자열 끝에 `\n` 을 직접 포함시켜야 합니다.

````python
# 'a' 모드: 앞서 만든 sample.txt 뒤에 이어 쓰기
with open("sample.txt", "a", encoding="utf-8") as f:
    f.write("네 번째 줄 (추가됨)\n")
    f.write("다섯 번째 줄 (추가됨)\n")

# 결과 확인: 기존 세 줄 + 추가된 두 줄 = 총 다섯 줄
with open("sample.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

````python
# writelines(): 리스트를 한 번에 파일에 씁니다
# 각 요소에 \n을 직접 포함시켜야 줄이 구분됩니다
lines = [
    "writelines 예제 첫 번째\n",
    "writelines 예제 두 번째\n",
    "writelines 예제 세 번째\n",
]

with open("writelines_test.txt", "w", encoding="utf-8") as f:
    f.writelines(lines)   # write()를 세 번 호출하는 것과 동일

# 결과 확인
with open("writelines_test.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

## 5. CSV 파일 기초

### CSV란?

CSV(Comma-Separated Values)는 쉼표로 각 열을 구분하는 텍스트 기반 표 형식입니다.
스프레드시트 소프트웨어(Excel, Google Sheets)와 호환되며, 데이터 교환에 가장 많이 사용되는 형식 중 하나입니다.

```
이름,나이,도시
홍길동,28,서울
김파이썬,35,부산
```

### 챕터 07 복습 — split()과 join() 활용

| 작업 | 코드 |
|------|------|
| CSV 한 줄 파싱 | `"홍길동,28,서울".split(",")` → `["홍길동", "28", "서울"]` |
| 리스트를 CSV 한 줄로 | `",".join(["홍길동", "28", "서울"])` → `"홍길동,28,서울"` |

기초 CSV는 `csv` 모듈 없이 `split()` / `join()` 만으로 처리할 수 있습니다.

### 실무에서는?

값 안에 쉼표나 줄바꿈이 포함된 복잡한 CSV는 파이썬 내장 `csv` 모듈을 사용합니다.
대용량 데이터 분석에는 `pandas` 라이브러리가 더 적합합니다.
두 도구 모두 이후 챕터에서 소개할 예정입니다.

````python
# CSV 파일 생성 (헤더 + 데이터 행)
header = ["이름", "나이", "도시"]
rows = [
    ["홍길동", "28", "서울"],
    ["김파이썬", "35", "부산"],
    ["이코딩", "22", "대전"],
]

with open("members.csv", "w", encoding="utf-8") as f:
    f.write(",".join(header) + "\n")
    for row in rows:
        f.write(",".join(row) + "\n")

print("members.csv 작성 완료")

# 파일 원본 내용 확인
with open("members.csv", "r", encoding="utf-8") as f:
    print(f.read())
````

````python
# CSV 파일 읽기 및 파싱
with open("members.csv", "r", encoding="utf-8") as f:
    lines = f.readlines()

# 첫 번째 줄은 헤더
header = lines[0].rstrip().split(",")
print(f"컬럼: {header}")
print()
````

````python

# 두 번째 줄부터 데이터
for line in lines[1:]:
    fields = line.rstrip().split(",")
    name, age, city = fields   # 언패킹으로 각 필드를 변수에 할당
    print(f"이름: {name}, 나이: {age}세, 도시: {city}")
````

## 챕터 정리

### 이번 챕터에서 배운 것

- **print() 심화**: `sep`으로 구분자 지정, `end`로 줄 끝 문자 변경, `file`로 출력 대상 파일 지정
- **input()**: 사용자 입력을 문자열로 받고, `int()` / `float()`으로 형변환
- **open() + with 문**: 파일을 안전하게 열고 자동으로 닫기
- **파일 모드**: `'r'`(읽기), `'w'`(쓰기/덮어쓰기), `'a'`(추가 쓰기)
- **읽기 함수**: `read()` 전체, `readlines()` 리스트, `readline()` 한 줄, `for` 순회
- **쓰기 함수**: `write()` 문자열 하나, `writelines()` 리스트
- **CSV 기초**: `split()` / `join()`으로 쉼표 구분 데이터 파싱 및 생성

### 핵심 함수 및 모드 정리

| 함수/모드 | 용도 |
|-----------|------|
| `print(..., sep=" ", end="\n", file=sys.stdout)` | 형식 지정 출력 |
| `input("안내 문구")` | 사용자 입력 (항상 str 반환) |
| `open(경로, 모드, encoding)` | 파일 열기 |
| `'r'` / `'w'` / `'a'` | 읽기 / 덮어쓰기 / 추가쓰기 |
| `f.read()` | 파일 전체를 문자열로 반환 |
| `f.readlines()` | 줄 단위 리스트 반환 |
| `f.readline()` | 한 줄씩 반환 (끝이면 `''`) |
| `f.write(str)` | 문자열 하나 쓰기 (`\n` 직접 포함) |
| `f.writelines(list)` | 리스트 한 번에 쓰기 |
