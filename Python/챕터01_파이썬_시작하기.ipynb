## 2. Hello, World! 작성

프로그래밍을 배울 때 가장 먼저 작성하는 코드는 전통적으로 "Hello, World!"입니다. 화면에 글자를 출력하는 가장 간단한 예제로, 환경이 제대로 설정되었는지 확인하는 용도이기도 합니다.

### print() 함수

화면에 내용을 출력할 때는 `print()` 함수를 사용합니다.

```
print("출력할 내용")
  |      |          |
 함수   따옴표     닫는 괄호
```

- `print`: 함수 이름입니다. "출력하라"는 명령입니다
- `(` `)`: 괄호 안에 출력할 내용을 넣습니다
- `"` `"`: 따옴표는 텍스트(문자열)임을 나타냅니다. 작은따옴표(`'`)도 사용 가능합니다

````python
print("Hello, World!")
````

`print()`는 한 번에 여러 값을 출력하거나, 여러 번 호출할 수도 있습니다. 숫자 계산 결과도 바로 출력할 수 있습니다.

````python
print("안녕하세요, Python!")
print("나의 첫 번째 프로그램입니다.")
print(1 + 1)           # 숫자도 출력 가능
print("결과:", 2 + 3) # 문자와 숫자 함께 출력
print("2 + 3")
````

## 3. Python 인터프리터 사용법 (REPL)

### REPL이란?

REPL은 다음 네 단계가 반복되는 대화형 실행 방식입니다.

```
Read   →   Eval   →   Print   →   Loop
읽고       실행하고    출력하고     반복
```

코드를 입력하면 즉시 실행되고 결과가 화면에 나타납니다. 파일을 저장하고 컴파일하는 과정 없이, 입력한 코드가 바로 실행됩니다.

### Jupyter Notebook과 REPL

지금 사용하고 있는 Jupyter Notebook이 바로 REPL 환경입니다. 셀에 코드를 입력하고 실행(`Shift + Enter`)하면 결과가 셀 아래에 즉시 표시됩니다.

**셀 하나 = REPL 한 사이클**

아래 셀들을 직접 실행해 보세요.

````python
# 셀에 수식을 입력하면 바로 결과가 출력됩니다
1 + 1
````

````python
10 * 5
````

````python
"Python" + " " + "공부 중"
````

### 주의: 마지막 줄만 자동 출력됩니다

셀 안에 여러 줄의 코드가 있을 때, 자동으로 출력되는 것은 **가장 마지막 줄의 결과**뿐입니다. 중간 결과를 확인하고 싶다면 `print()`를 사용해야 합니다.

````python
# 여러 줄일 때는 마지막 줄만 자동으로 표시됩니다
100 + 200
300 + 400   # 이 줄의 결과만 출력됨
````

````python
# 중간 결과도 보고 싶다면 print()를 사용하세요
print(100 + 200)
print(300 + 400)
````

## 4. 주석 작성법

### 주석이란?

주석(Comment)은 코드에 달아두는 메모입니다. Python은 주석을 코드로 인식하지 않으며, 실행할 때 완전히 무시합니다.

### 주석이 필요한 이유

- 코드가 **왜** 이렇게 작성되었는지 설명하기 위해
- 나중에 코드를 다시 읽을 때 빠르게 이해하기 위해
- 팀원과 협업할 때 의도를 전달하기 위해

> 좋은 주석은 "무엇을 하는 코드"가 아니라 "왜 이렇게 했는가"를 설명합니다.

### 두 가지 주석 방법

**한 줄 주석**: `#` 기호를 사용합니다

```python
# 이것은 한 줄 주석입니다
print("코드")  # 코드 뒤에도 달 수 있습니다
```

**여러 줄 주석**: 따옴표 세 개(`"""` 또는 `'''`)로 감쌉니다

```python
"""
여러 줄에 걸친
긴 설명을 작성할 때 사용합니다
"""
```

````python
# 이 줄은 주석입니다. 실행되지 않습니다.
print("이 줄은 실행됩니다.")  # 코드 뒤에도 주석을 달 수 있습니다

# 주석은 코드의 '왜'를 설명하는 데 사용합니다
# 아래 코드는 두 숫자를 더합니다
result = 10 + 20
print(result)
````

````python
"""
이것은 여러 줄 주석입니다.
긴 설명이 필요할 때 사용합니다.
따옴표 세 개로 시작하고 세 개로 끝납니다.
"""
print("여러 줄 주석 아래의 코드는 정상 실행됩니다.")
````
