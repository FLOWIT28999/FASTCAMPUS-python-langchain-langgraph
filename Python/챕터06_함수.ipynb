## 1. 함수란 무엇인가

**함수(function)** 는 특정 작업을 수행하는 코드 블록에 이름을 붙인 것이다.  
한 번 정의해 두면 이름만 호출해서 언제든지 재사용할 수 있다.

### 왜 함수를 쓰는가

| 이유 | 설명 |
|------|------|
| **중복 제거** | 같은 로직을 여러 곳에 반복해서 작성하지 않아도 된다 |
| **가독성 향상** | 코드가 무슨 일을 하는지 함수 이름만 봐도 파악할 수 있다 |
| **유지보수 용이** | 로직을 수정할 때 함수 내부만 바꾸면 호출 부분은 그대로 유지된다 |
| **테스트 편의** | 함수 단위로 독립적으로 테스트할 수 있다 |

아래 예제에서 함수가 없을 때와 있을 때의 차이를 직접 확인해 보자.

````python
# 함수 없이: 같은 로직을 매번 반복해야 한다
print("=" * 20)
print("결과 출력")
print("=" * 20)

print("=" * 20)
print("두 번째 출력")
print("=" * 20)

print()  # 구분용 빈 줄
````

````python

# 함수로 묶으면: 한 번 정의하고 여러 번 호출한다
def print_box(message):
    print("=" * 20)
    print(message)
    print("=" * 20)

print_box("결과 출력")
print_box("두 번째 출력")
````

## 2. 함수 정의와 호출 / return

### 기본 구조

```
def 함수이름(매개변수):
    실행할 코드
    return 반환값
```

### 핵심 용어

| 용어 | 설명 | 예시 |
|------|------|------|
| **매개변수 (parameter)** | 함수를 **정의**할 때 사용하는 변수 이름 | `def greet(name):` 의 `name` |
| **인자 (argument)** | 함수를 **호출**할 때 실제로 전달하는 값 | `greet("김파이썬")` 의 `"김파이썬"` |
| **return** | 함수의 결과를 호출한 곳으로 돌려준다 | `return result` |

> **주의**: `return` 문이 없으면 함수는 자동으로 `None`을 반환한다.

````python
# 기본 함수: 정의 → 호출 → return
def greet(name):
    return f"안녕하세요, {name}님!"

result = greet("김파이썬")
print(result)

print()

# 여러 값을 return하면 튜플로 묶여 반환된다
def min_max(numbers):
    return min(numbers), max(numbers)

lo, hi = min_max([3, 1, 7, 2, 9])
print(f"최솟값: {lo}, 최댓값: {hi}")
````

````python
# return이 없는 함수는 None을 반환한다
def print_line(char="-", length=20):
    print(char * length)

result = print_line()
print("반환값:", result)        # None
print("타입:", type(result))    # <class 'NoneType'>
````

## 3. 기본값 매개변수 / 키워드 인자

### 기본값 매개변수 (default parameter)

매개변수에 기본값을 지정하면, 호출 시 해당 인자를 생략해도 기본값이 사용된다.

```python
def power(base, exp=2):   # exp의 기본값은 2
    return base ** exp
```

> **규칙**: 기본값이 있는 매개변수는 반드시 기본값이 없는 매개변수 **뒤에** 위치해야 한다.  
> `def foo(a=1, b)` 는 오류 — `def foo(b, a=1)` 처럼 작성한다.

### 키워드 인자 (keyword argument)

함수를 호출할 때 `이름=값` 형태로 인자를 전달하면 순서에 관계없이 원하는 매개변수에 값을 넘길 수 있다.

````python
# 기본값 매개변수
def power(base, exp=2):
    return base ** exp

print(power(3))       # 9  — exp 기본값 2 사용
print(power(3, 3))    # 27 — exp를 3으로 지정
print(power(2, 10))   # 1024
````

````python
# 키워드 인자: 이름을 명시하면 순서에 관계없이 전달 가능
def introduce(name, age, city):
    print(f"이름: {name}, 나이: {age}, 도시: {city}")

introduce("김파이썬", 25, "서울")                   # 위치 인자: 순서대로
introduce(age=30, city="부산", name="이파이썬")      # 키워드 인자: 순서 무관
````

## 4. 가변 인자 (*args / **kwargs)

인자의 개수가 미리 정해지지 않을 때 사용한다.

| 형태 | 의미 | 함수 내부 타입 |
|------|------|----------------|
| `*args` | 위치 인자를 **여러 개** 받음 | 튜플 (`tuple`) |
| `**kwargs` | 키워드 인자를 **여러 개** 받음 | 딕셔너리 (`dict`) |

관례적으로 `args`, `kwargs`라는 이름을 사용하지만, 별표(`*`, `**`)가 핵심이므로 다른 이름도 사용 가능하다.

````python
# *args: 위치 인자를 튜플로 받는다
def total(*args):
    result = 0
    for n in args:
        result += n
    return result

print(total(1, 2, 3))           # 6
print(total(10, 20, 30, 40))    # 100
print(total())               # 0 — 인자 없음도 허용
````

````python
# **kwargs: 키워드 인자를 딕셔너리로 받는다
def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

show_info(name="김파이썬", age=25, city="서울")
print()
show_info(language="Python", version=3.12)
````

## 5. 람다 함수 (lambda)

`lambda`는 이름 없이 한 줄로 정의하는 **익명 함수**다.

### 구문

```
lambda 매개변수: 표현식
```

### 언제 쓰는가

- `sorted()`, `filter()`, `map()` 등 **함수를 인자로 받는 함수**에 짧은 함수를 인라인으로 전달할 때
- 한 번만 사용할 간단한 로직을 별도 `def`로 정의하기 번거로울 때

> 람다 안에서는 표현식(expression)만 가능하다 — `if/for` 같은 구문(statement)은 사용할 수 없다.

````python
# 일반 함수와 람다 비교
def square_fn(x):
    return x ** 2

square_lambda = lambda x: x ** 2

print(square_fn(5))       # 25
print(square_lambda(5))   # 25

print()

# 두 인자를 받는 람다
add = lambda a, b: a + b
print(add(3, 7))   # 10
````

````python
# sorted와 람다: 정렬 기준을 key 인자로 지정한다
students = [("김파이썬", 85), ("이자바", 92), ("박스크립트", 78)]

# 점수 오름차순 정렬
sorted_asc = sorted(students, key=lambda s: s[1])
print("점수 오름차순:")
for name, score in sorted_asc:
    print(f"  {name}: {score}")

print()

# 점수 내림차순 정렬
sorted_desc = sorted(students, key=lambda s: s[1], reverse=True)
print("점수 내림차순:")
for name, score in sorted_desc:
    print(f"  {name}: {score}")
````

## 6. 스코프 (Scope)

**스코프(scope)** 는 변수가 유효한 범위다.  
파이썬은 변수를 탐색할 때 **LEGB 규칙** 에 따라 안쪽에서 바깥쪽으로 순서대로 찾는다.

| 범위 | 영문 | 설명 |
|------|------|------|
| **L** | Local | 현재 함수 안 |
| **E** | Enclosing | 중첩 함수의 바깥 함수 |
| **G** | Global | 모듈(파일) 최상위 |
| **B** | Built-in | `print`, `len` 등 파이썬 내장 이름 |

이 챕터에서는 **Local**과 **Global**에 집중한다.

### global 키워드

함수 안에서 전역 변수의 값을 **수정**하려면 `global` 키워드로 먼저 선언해야 한다.  
선언 없이 값을 바꾸려 하면 지역 변수로 새로 만들어지거나 `UnboundLocalError`가 발생한다.

````python
# 지역 변수 vs 전역 변수
x = 10  # 전역 변수 — 파일 최상위에서 정의

def show_scope():
    y = 20  # 지역 변수 — 이 함수 안에서만 유효
    print("함수 안 x:", x)   # 전역 변수 읽기는 가능
    print("함수 안 y:", y)

show_scope()
print("함수 밖 x:", x)
# print(y)  # 주석을 풀면 NameError 발생 — y는 함수 밖에서 접근할 수 없다
````

````python
# global 키워드: 함수 안에서 전역 변수를 수정한다
count = 0  # 전역 변수

def increment():
    global count   # count가 전역 변수임을 선언
    count += 1

increment()
increment()
increment()
print("count:", count)  # 3
````

## 챕터 정리

### 이번 챕터에서 배운 내용

- `def 함수이름(매개변수):` 형태로 함수를 정의하고, 이름 뒤에 `()`를 붙여 호출한다.
- `return`으로 결과를 돌려주며, `return`이 없으면 `None`이 반환된다.
- 매개변수(parameter)는 정의할 때 쓰는 이름이고, 인자(argument)는 호출할 때 전달하는 값이다.
- 기본값 매개변수로 인자 생략 시 사용할 값을 지정하고, 키워드 인자로 순서에 무관하게 전달한다.
- `*args`는 위치 인자를 튜플로, `**kwargs`는 키워드 인자를 딕셔너리로 묶어 받는다.
- `lambda 매개변수: 표현식`으로 이름 없는 한 줄 함수를 만들 수 있다.
- 변수의 유효 범위(스코프)는 LEGB 규칙으로 탐색되며, 전역 변수를 함수 안에서 수정하려면 `global`을 선언한다.

### 핵심 키워드 한눈에 보기

| 분류 | 키워드/문법 | 역할 |
|------|-------------|------|
| 함수 정의 | `def` | 함수 블록 시작 |
| 반환 | `return` | 결과를 호출부로 돌려줌 |
| 기본값 | `def f(x=값)` | 인자 생략 시 사용할 기본값 지정 |
| 키워드 인자 | `f(이름=값)` | 순서에 무관하게 인자 전달 |
| 가변 위치 인자 | `*args` | 위치 인자 여러 개를 튜플로 수집 |
| 가변 키워드 인자 | `**kwargs` | 키워드 인자 여러 개를 딕셔너리로 수집 |
| 익명 함수 | `lambda` | 이름 없는 한 줄 함수 |
| 전역 수정 | `global` | 함수 안에서 전역 변수 수정 선언 |
