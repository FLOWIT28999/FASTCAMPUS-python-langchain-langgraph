## 이 챕터에서 배울 내용

## 1. 산술 연산자

산술 연산자는 **수학 계산**을 수행합니다.

| 연산자 | 이름 | 설명 | 예시 | 결과 |
|--------|------|------|------|------|
| `+` | 덧셈 | 두 값을 더한다 | `10 + 3` | `13` |
| `-` | 뺄셈 | 왼쪽에서 오른쪽을 뺀다 | `10 - 3` | `7` |
| `*` | 곱셈 | 두 값을 곱한다 | `10 * 3` | `30` |
| `/` | 나누기 | 나눈 결과를 실수로 반환 | `10 / 3` | `3.333...` |
| `//` | 정수 나누기 | 나눈 결과의 몫(정수)만 반환 | `10 // 3` | `3` |
| `%` | 나머지 | 나눈 결과의 나머지를 반환 | `10 % 3` | `1` |
| `**` | 거듭제곱 | 왼쪽을 오른쪽만큼 제곱 | `2 ** 8` | `256` |

### `/` vs `//` 구분

챕터 02에서 짧게 언급했던 내용입니다.

- `/` : 결과가 항상 **float**입니다. `10 / 2 = 5.0`
- `//` : 소수점 이하를 버리고 **정수 몫**만 반환합니다. `10 // 3 = 3`

````python
# 기본 사칙연산
print(10 + 3)  # 13
print(10 - 3)  # 7
print(10 * 3)  # 30
print(10 / 3)  # 3.3333... (항상 float 반환)
````

````python
# 나머지, 몫, 거듭제곱
print(10 // 3)  # 몫: 3
print(10 % 3)   # 나머지: 1
print(2 ** 8)   # 거듭제곱: 256
````

````python
# 나머지로 홀짝 확인
number = 17
print(number % 2)  # 1이면 홀수, 0이면 짝수

# 총금액 계산
price = 1500
quantity = 4
total = price * quantity
print("총 금액:", total, "원")
````

## 2. 비교 연산자

비교 연산자는 두 값을 비교해 **항상 `True` 또는 `False`를 반환**합니다.

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `==` | 같다 | `10 == 10` | `True` |
| `!=` | 다르다 | `10 != 5` | `True` |
| `>` | 크다 | `10 > 5` | `True` |
| `<` | 작다 | `10 < 5` | `False` |
| `>=` | 크거나 같다 | `10 >= 10` | `True` |
| `<=` | 작거나 같다 | `10 <= 9` | `False` |

### 주의: `=`(할당)과 `==`(비교) 혼동 금지

```python
x = 10    # 할당: x에 10을 저장한다
x == 10   # 비교: x가 10과 같은가? → True 또는 False
```

조건을 확인할 때 `=`를 쓰면 오류가 발생합니다. 반드시 `==`를 사용하세요.

````python
# 기본 비교
print(10 == 10)  # True
print(10 != 5)   # True
print(10 > 5)    # True
print(10 < 5)    # False
print(10 >= 10)  # True
print(10 <= 9)   # False
````

````python
# 변수끼리 비교
a = 85
b = 90
print("a가 b보다 큰가?", a > b)
print("a와 b가 같은가?", a == b)
print("a가 합격 점수(80점) 이상인가?", a >= 80)
````

## 3. 논리 연산자

논리 연산자는 **여러 조건을 조합**할 때 사용합니다.

| 연산자 | 의미 | 설명 |
|--------|------|------|
| `and` | 그리고 | 둘 다 `True`여야 `True` |
| `or` | 또는 | 하나라도 `True`면 `True` |
| `not` | 부정 | `True` ↔ `False` 반전 |

### 진리표 (Truth Table)

| A | B | A and B | A or B | not A |
|---|---|---------|--------|-------|
| `True` | `True` | `True` | `True` | `False` |
| `True` | `False` | `False` | `True` | `False` |
| `False` | `True` | `False` | `True` | `True` |
| `False` | `False` | `False` | `False` | `True` |

````python
# 기본 논리 연산
print(True and True)   # True
print(True and False)  # False
print(True or False)   # True
print(False or False)  # False
print(not True)        # False
print(not False)       # True
````

````python
# 조건 조합 예시
age = 20
has_ticket = True

# 입장 가능 조건: 18세 이상 AND 티켓 보유
can_enter = age >= 18 and has_ticket 
print("입장 가능:", can_enter)

# 할인 조건: 어린이(13세 미만) OR 노인(65세 이상)
age2 = 10
is_discounted = age2 < 13 or age2 >= 65
print("할인 적용:", is_discounted)
````

## 4. 할당 연산자

기본 할당 `=` 외에, **계산과 할당을 한 번에** 처리하는 복합 할당 연산자가 있습니다.

| 연산자 | 의미 | 동일한 표현 |
|--------|------|-------------|
| `=` | 할당 | `x = 5` |
| `+=` | 더하고 할당 | `x += 1` → `x = x + 1` |
| `-=` | 빼고 할당 | `x -= 1` → `x = x - 1` |
| `*=` | 곱하고 할당 | `x *= 2` → `x = x * 2` |
| `/=` | 나누고 할당 | `x /= 2` → `x = x / 2` |
| `//=` | 몫을 할당 | `x //= 3` → `x = x // 3` |
| `%=` | 나머지를 할당 | `x %= 3` → `x = x % 3` |
| `**=` | 거듭제곱 후 할당 | `x **= 2` → `x = x ** 2` |

`x += 1`은 `x = x + 1`과 완전히 동일합니다. 코드를 더 간결하게 쓸 수 있습니다.

````python
score = 100
print("초기:", score)

score += 10   # score = score + 10
print("+10 후:", score)

score -= 5    # score = score - 5
print("-5 후:", score)

score *= 2    # score = score * 2
print("x2 후:", score)

score //= 3   # score = score // 3
print("//3 후:", score)

print(score)
````

## 5. 멤버십 연산자

멤버십 연산자는 **어떤 값이 컨테이너(문자열, 리스트 등)에 포함되어 있는지** 확인합니다.

| 연산자 | 의미 | 설명 |
|--------|------|------|
| `in` | 포함됨 | 왼쪽 값이 오른쪽 컨테이너에 있으면 `True` |
| `not in` | 포함되지 않음 | 왼쪽 값이 오른쪽 컨테이너에 없으면 `True` |

문자열, 리스트, 튜플 등 다양한 컨테이너에 사용할 수 있습니다.

````python
# 문자열에서 in 사용
message = "파이썬은 재미있습니다"
print("파이썬" in message)     # True
print("자바" in message)       # False
print("자바" not in message)   # True

# 리스트에서 in 사용
fruits = ["사과", "바나나", "포도"]
print("바나나" in fruits)      # True
print("딸기" not in fruits)    # True
````

## 6. 식별 연산자

식별 연산자는 두 변수가 **같은 객체(메모리 주소)를 가리키는지** 확인합니다.

| 연산자 | 의미 |
|--------|------|
| `is` | 같은 객체면 `True` |
| `is not` | 다른 객체면 `True` |

### `==`(값이 같다) vs `is`(같은 객체다)

```python
a = [1, 2, 3]
b = [1, 2, 3]

a == b   # True  — 값이 같다
a is b   # False — 메모리 주소는 다른 별개의 리스트
```

### 실용적 사용처

일반적으로 값 비교에는 `==`를 사용합니다. `is`는 주로 **`None`인지 확인**할 때 사용하는 것이 관례입니다.

```python
# 권장하는 방법
if result is None:
    ...
```

````python
# None 확인에 is 사용 (권장)
result = None
print(result is None)      # True
print(result is not None)  # False

# == 와 is의 차이
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a == b)   # True  (값이 같다)
print(a is b)   # False (다른 객체)
print(a is c)   # True  (같은 객체를 가리킴)
````

## 7. 연산자 우선순위

수식에 여러 연산자가 섞여 있을 때, **어떤 연산을 먼저 수행할지**를 정하는 규칙입니다.

수학과 동일한 원칙을 따릅니다.

| 우선순위 | 연산자 | 설명 |
|----------|--------|------|
| 1 (가장 높음) | `()` | 괄호 |
| 2 | `**` | 거듭제곱 |
| 3 | `+x`, `-x` | 단항 부호 |
| 4 | `*`, `/`, `//`, `%` | 곱셈, 나눗셈 |
| 5 | `+`, `-` | 덧셈, 뺄셈 |
| 6 | `==`, `!=`, `>`, `<`, `>=`, `<=`, `in`, `is` | 비교 / 멤버십 / 식별 |
| 7 | `not` | 논리 부정 |
| 8 | `and` | 논리 곱 |
| 9 (가장 낮음) | `or` | 논리 합 |

우선순위가 헷갈릴 때는 **괄호를 사용**하는 것이 가장 안전하고 명확합니다.

````python
# 우선순위에 따른 계산 결과 차이
print(2 + 3 * 4)    # 14  (곱셈 먼저)
print((2 + 3) * 4)  # 20  (괄호 먼저)

print(2 ** 3 ** 2)  # 512 (거듭제곱은 오른쪽부터: 3**2=9, 2**9=512)
print(10 - 4 - 2)   # 4   (같은 우선순위는 왼쪽부터)
````

````python
# 논리 연산자 우선순위: not > and > or
print(True or False and False)    # True  (and 먼저: False and False=False, True or False=True)
print((True or False) and False)  # False (괄호 먼저: True and False=False)
print(not True or True)           # True  (not 먼저: False or True=True)
````

## 챕터 정리

이번 챕터에서 배운 내용을 정리합니다.

- **산술 연산자**: `+`, `-`, `*`, `/`, `//`, `%`, `**`로 수학 계산을 수행합니다. `/`는 항상 float, `//`는 정수 몫을 반환합니다
- **비교 연산자**: `==`, `!=`, `>`, `<`, `>=`, `<=`의 결과는 항상 `True`/`False`입니다. `=`(할당)과 `==`(비교)를 혼동하지 않습니다
- **논리 연산자**: `and`(둘 다 참), `or`(하나라도 참), `not`(반전)으로 여러 조건을 조합합니다
- **할당 연산자**: `+=`, `-=` 등 복합 할당 연산자로 계산과 저장을 한 번에 처리합니다
- **멤버십 연산자**: `in`, `not in`으로 값이 컨테이너에 포함되어 있는지 확인합니다
- **식별 연산자**: `is`, `is not`으로 같은 객체인지 확인합니다. 주로 `None` 확인에 사용합니다
- **연산자 우선순위**: 괄호 > 거듭제곱 > 곱셈·나눗셈 > 덧셈·뺄셈 > 비교 > not > and > or. 헷갈리면 괄호로 명확히 합니다

### 연산자 한눈에 비교

| 종류 | 연산자 | 반환 타입 |
|------|--------|----------|
| 산술 | `+` `-` `*` `/` `//` `%` `**` | 숫자 (`int` 또는 `float`) |
| 비교 | `==` `!=` `>` `<` `>=` `<=` | `bool` |
| 논리 | `and` `or` `not` | `bool` |
| 할당 | `=` `+=` `-=` `*=` `/=` `//=` `%=` `**=` | (없음, 변수에 저장) |
| 멤버십 | `in` `not in` | `bool` |
| 식별 | `is` `is not` | `bool` |
