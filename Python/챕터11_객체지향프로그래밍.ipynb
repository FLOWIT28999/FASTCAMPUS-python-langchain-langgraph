## 1. 클래스와 객체

### 왜 클래스가 필요한가?

클래스 없이 강아지 여러 마리를 관리한다고 해보자.

```python
dog1_name, dog1_age = "Luna", 3
dog2_name, dog2_age = "Rex",  5
dog3_name, dog3_age = "Max",  2
```

강아지가 10마리면 변수 20개, 출력 코드도 10번 반복된다.  
`dog2_name`을 수정하다가 `dog3_name`을 실수로 건드려도 에러가 없다.  
→ **클래스**는 관련 데이터(이름, 나이)와 동작(짖기, 소개)을 하나의 단위로 묶어 이 문제를 해결한다.

### 객체지향 프로그래밍(OOP)이란?

**데이터(속성)**와 **동작(메서드)**을 하나의 단위로 묶어 프로그램을 설계하는 방식이다.  
현실 세계의 사물을 코드로 모델링할 때 직관적으로 표현할 수 있다.

- **클래스(Class)**: 사물의 설계도. 어떤 속성과 동작을 가질지 정의한다.
- **인스턴스(Instance)**: 설계도로 만들어 낸 실제 사물. 각각 독립된 데이터를 갖는다.

```
클래스 = 붕어빵 틀
인스턴스 = 틀로 찍어낸 붕어빵 (팥맛, 크림맛 — 각각 다른 속성)
```

> **`self`란?** 메서드를 호출하면 Python이 자동으로 그 인스턴스를 첫 번째 인자로 넘겨준다.  
> `luna.bark()` 를 호출하면 내부적으로 `Dog.bark(luna)` 와 동일하게 처리된다.  
> 관례적으로 이 첫 번째 인자를 `self`라고 이름 붙인다 — 직접 넘기지 않아도 Python이 자동으로 처리한다.

### 클래스 변수 vs 인스턴스 변수

| 종류 | 위치 | 공유 여부 | 접근 방법 |
|------|------|----------|-----------|
| 클래스 변수 | 클래스 블록, 메서드 밖 | 모든 인스턴스가 공유 | `클래스명.변수` 또는 `self.변수` |
| 인스턴스 변수 | `__init__` 안, `self.변수` | 각 인스턴스가 독립 보유 | `self.변수` |

### 기본 문법

```python
class 클래스명:
    클래스변수 = 값            # 모든 인스턴스가 공유

    def __init__(self, ...):  # 생성자: 인스턴스 생성 시 자동 실행
        self.인스턴스변수 = 값  # 인스턴스마다 고유한 변수

    def 메서드(self):         # 인스턴스 메서드
        ...
```

````python
# ── 클래스 없이 강아지를 관리하면? ───────────────────────────────────────
dog1_name, dog1_age = "Luna", 3
dog2_name, dog2_age = "Rex",  5
print(f"Name: {dog1_name}, Age: {dog1_age}")
print(f"Name: {dog2_name}, Age: {dog2_age}")
print("→ 강아지가 10마리면 변수 20개, 출력 코드도 10번 반복됨")
print()

# ── Step 1: 최소한의 클래스 — __init__만 있는 버전 ───────────────────────
# __init__ : 인스턴스가 만들어질 때 자동으로 실행되는 메서드 (생성자)
# self     : Python이 자동으로 넘겨주는 "이 인스턴스 자신"에 대한 참조
#            luna = Dog("Luna", 3) 호출 시 → self 자리에 luna가 자동으로 들어옴
class Dog:
    def __init__(self, name, age):
        self.name = name    # self.변수 → 이 인스턴스만의 데이터로 저장
        self.age  = age     # luna.name = "Luna", rex.name = "Rex" — 각자 독립

luna = Dog("Luna", 3)
rex  = Dog("Rex",  5)

print(f"{luna.name}, {luna.age}살")   # 각 인스턴스가 독립적으로 데이터 보유
print(f"{rex.name},  {rex.age}살")
````

````python
# ── Step 2: 관련 동작(메서드)을 클래스 안에 추가 ─────────────────────────
# luna.introduce() 를 호출하면 Python은 Dog.introduce(luna) 처럼 처리한다
# 즉 self에 luna가 자동으로 들어온다 — 호출할 때 self를 직접 넘기지 않아도 된다
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def introduce(self):              # self 를 통해 이 인스턴스의 데이터에 접근
        return f"Name: {self.name}, Age: {self.age}"

    def bark(self):
        return f"{self.name}: Woof!"

luna = Dog("Luna", 3)
print(luna.introduce())   # Python이 Dog.introduce(luna) 로 처리
print(luna.bark())
print()
````

````python

# ── Step 3: 모든 인스턴스가 공유하는 값은 클래스 변수로 ─────────────────
# 인스턴스 변수 → 인스턴스마다 다른 값 (이름, 나이)
# 클래스 변수  → 모든 인스턴스가 같은 값을 공유 (종)
class Dog:
    species = "Canis familiaris"     # 클래스 변수 — Dog 클래스 자체에 속함

    def __init__(self, name, age):
        self.name = name             # 인스턴스 변수 — 각 강아지마다 다름
        self.age  = age

    def introduce(self):
        return f"Name: {self.name}, Age: {self.age}, Species: {self.species}"

luna = Dog("Luna", 3)
rex  = Dog("Rex",  5)

print(luna.introduce())
print(rex.introduce())
print()
print(f"Dog.species  = {Dog.species}")     # 클래스로 직접 접근
print(f"luna.species = {luna.species}")    # 인스턴스를 통해 접근 — 같은 값
print()
````

````python

# ── Student 클래스로 더 연습 ──────────────────────────────────────────────
class Student:
    school = "Python Academy"   # 클래스 변수

    def __init__(self, name, grade_level, scores):
        self.name        = name
        self.grade_level = grade_level
        self.scores      = scores

    def average(self):
        return sum(self.scores) / len(self.scores)

    def letter_grade(self):
        avg = self.average()
        if avg >= 90:   return "A"
        elif avg >= 80: return "B"
        elif avg >= 70: return "C"
        else:           return "D"

    def print_info(self):
        print(f"[{self.school}] Grade {self.grade_level} — {self.name}")
        print(f"  Scores: {self.scores}")
        print(f"  Average: {self.average():.1f}  Grade: {self.letter_grade()}")

student1 = Student("Alice", 1, [85, 92, 78, 95, 88])
student2 = Student("Bob",   2, [92, 96, 89, 94, 91])

student1.print_info()
print()
student2.print_info()
````

## 2. 메서드 3종류

### 왜 3종류가 필요한가?

인스턴스 메서드(`self`) 하나로 모든 걸 해결하려 하면 안 되는 상황이 있다.

**상황 1 — 인스턴스가 아직 없는데 객체를 만들고 싶다**  
`Circle.from_diameter(10)` 처럼 지름만 알고 있을 때 원을 만들고 싶다.  
아직 인스턴스가 없으니 `self`를 쓸 수 없다 → **`@classmethod`** 가 필요하다.

**상황 2 — 클래스와 논리적으로 관련은 있지만, 아무 인스턴스·클래스 데이터도 필요 없다**  
`5cm → 50mm` 단위 변환처럼 `Circle` 클래스와 관련은 있지만 self/cls가 필요 없다  
→ **`@staticmethod`** 가 필요하다.

| 종류 | 데코레이터 | 첫 매개변수 | 접근 가능한 것 | 용도 |
|------|-----------|------------|--------------|------|
| 인스턴스 메서드 | 없음 | `self` | 인스턴스 변수, 클래스 변수 | 인스턴스 데이터 처리 |
| 클래스 메서드 | `@classmethod` | `cls` | 클래스 변수 | 대체 생성자, 클래스 수준 기능 |
| 정적 메서드 | `@staticmethod` | 없음 | 없음 | 클래스와 관련된 독립 유틸리티 |

- **`self`**: 인스턴스 자신을 가리키는 관례적 이름
- **`cls`**: 클래스 자신을 가리키는 관례적 이름
- **`@classmethod`**: 인스턴스 없이 클래스 이름으로 호출, 대체 생성자 패턴에 자주 사용
- **`@staticmethod`**: 클래스·인스턴스와 무관한 독립 함수, 관련 유틸리티를 클래스 안에 묶을 때 사용

````python
# ── Step 1: 인스턴스 메서드만 있는 Circle ────────────────────────────────
class Circle:
    pi = 3.14159

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.pi * self.radius ** 2

    def perimeter(self):
        return 2 * Circle.pi * self.radius

c1 = Circle(5)
print(f"반지름 5: 넓이={c1.area():.2f}, 둘레={c1.perimeter():.2f}")
print()

# ── 문제: 지름(10)만 알고 있을 때 Circle을 만들고 싶다면? ────────────────
# 방법 A: 호출하는 쪽에서 매번 직접 나눠야 한다 → 실수하기 쉽다
c2 = Circle(10 / 2)   # "지름으로 만든다"는 의도가 코드에 드러나지 않는다
print(f"방법 A (직접 나누기): {c2.radius}")

# ── Step 2: @classmethod — 지름을 받는 대체 생성자 ──────────────────────
# cls 는 Circle 클래스 자체를 가리킨다
# cls(diameter / 2) 는 Circle(diameter / 2) 와 동일
class Circle:
    pi = 3.14159

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.pi * self.radius ** 2

    def perimeter(self):
        return 2 * Circle.pi * self.radius

    @classmethod
    def from_diameter(cls, diameter):   # 인스턴스 없이 클래스로 호출
        return cls(diameter / 2)        # cls() = Circle() — 서브클래스에서도 작동

c3 = Circle.from_diameter(10)   # 의도가 명확하다
print(f"방법 B (@classmethod):  {c3.radius}")
print()

# ── Step 3: @staticmethod — 인스턴스/클래스 데이터가 필요 없는 유틸리티 ──
class Circle:
    pi = 3.14159

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return Circle.pi * self.radius ** 2

    def perimeter(self):
        return 2 * Circle.pi * self.radius

    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)

    @staticmethod
    def cm_to_mm(cm):       # self도 cls도 필요 없다 — 순수 유틸리티
        return cm * 10

    def __str__(self):
        return f"Circle(radius={self.radius})"

c1 = Circle(5)
print(f"{c1}: area={c1.area():.2f}, perimeter={c1.perimeter():.2f}")
c2 = Circle.from_diameter(10)
print(f"{c2}: area={c2.area():.2f}")
print(f"5 cm = {Circle.cm_to_mm(5)} mm   ← 인스턴스 없이 바로 호출 가능")
````

## 3. 상속

### 왜 상속이 필요한가?

고양이(`Cat`)와 강아지(`Dog`) 클래스를 따로 만들면 어떻게 될까?

```python
class Cat:
    def __init__(self, name, age):   # ← Dog와 완전히 같은 코드
        self.name = name
        self.age  = age
    def introduce(self):             # ← Dog와 거의 같은 코드 → 중복!
        return f"{self.name} ({self.age} years old)"

class Dog:
    def __init__(self, name, age):   # ← 중복
        self.name = name
        self.age  = age
    def introduce(self):             # ← 중복
        return f"{self.name} ({self.age} years old)"
```

동물 클래스가 10개라면 `__init__`이 10번, `introduce()`가 10번 중복된다.  
`introduce()` 형식을 바꾸려면 10개 클래스를 모두 수정해야 한다.

→ **상속**을 쓰면 공통 코드는 부모(`Animal`)에 한 번만 작성하고,  
자식(`Cat`, `Dog`)은 자신만의 부분만 추가하거나 바꾸면 된다.

### 핵심 개념

- **`super()`**: 자식 클래스에서 부모 클래스의 메서드를 호출
- **메서드 오버라이딩**: 부모의 메서드를 자식에서 같은 이름으로 재정의
- **`isinstance(객체, 클래스)`**: 객체가 해당 클래스(또는 자식 클래스)의 인스턴스인지 확인
- **`issubclass(자식, 부모)`**: 클래스 간 상속 관계 확인

> **`super().__init__()` 를 항상 먼저 호출하자.**  
> 자식의 `__init__` 에서 `super().__init__()` 를 빠뜨리면 부모가 설정하는 `self.name` 같은 속성이 존재하지 않아 `AttributeError` 가 발생한다.

```python
class 자식클래스(부모클래스):
    def __init__(self, ...):
        super().__init__(...)   # 부모 생성자 먼저 호출 → self.name 등이 설정됨
        self.추가속성 = 값       # 그 다음에 자식 고유 속성 추가

    def 오버라이딩할메서드(self):  # 부모 메서드 재정의
        ...
```

````python
# ── Before: 상속 없이 Cat과 Dog를 따로 정의하면? ─────────────────────────
class CatNoInheritance:
    def __init__(self, name, age, breed):
        self.name  = name    # ↓ Dog와 완전히 같은 코드가 그대로 중복됨
        self.age   = age
        self.breed = breed

    def introduce(self):     # Dog의 introduce()와 거의 같다 → 중복!
        return f"{self.name} ({self.age} years old) [{self.breed}]"

    def make_sound(self):
        return "Meow!"

class DogNoInheritance:
    def __init__(self, name, age):
        self.name = name     # ← Cat과 중복
        self.age  = age      # ← 중복

    def introduce(self):     # ← 거의 같은 코드를 또 작성
        return f"{self.name} ({self.age} years old)"

    def make_sound(self):
        return "Woof!"

cat = CatNoInheritance("Luna", 2, "Persian")
dog = DogNoInheritance("Rex",  4)
print(cat.introduce(), "—", cat.make_sound())
print(dog.introduce(), "—", dog.make_sound())
print()
print("→ introduce() 형식을 바꾸려면 두 클래스를 모두 수정해야 한다.")
print("  동물 종류가 10개면? 같은 수정을 10번 반복해야 한다.")
print("  다음 셀에서 상속으로 이 중복을 제거한다.")
````

````python
class Animal:    # 부모 클래스 — 모든 동물의 공통 속성과 동작 정의
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def introduce(self):
        return f"{self.name} ({self.age} years old)"

    def make_sound(self):
        return "..."   # 기본 소리 — 자식 클래스에서 오버라이딩

class Cat(Animal):   # 자식 클래스 — name, age, introduce(), make_sound() 상속
    def __init__(self, name, age, breed):
        super().__init__(name, age)   # Animal.__init__ 호출 → self.name, self.age 설정
        self.breed = breed            # Cat만의 고유 속성 추가

    def make_sound(self):             # 오버라이딩 — Cat 고유의 소리
        return "Meow!"

    def introduce(self):              # 확장 — 부모 버전 호출 후 품종 정보 추가
        return f"{super().introduce()} [{self.breed}]"

# 주의: Dog 클래스는 5절(다형성)에서 같은 이름으로 재정의됩니다.
# 이 버전은 상속 개념의 단독 예제입니다.
class Dog(Animal):
    def make_sound(self):             # 오버라이딩 — Dog 고유의 소리
        return "Woof!"

    def perform_trick(self, action):
        return f"{self.name} performs '{action}'!"

# 인스턴스 생성
luna = Cat("Luna", 2, "Persian")
rex  = Dog("Rex", 4)

print(luna.introduce())
print(rex.introduce())   # Animal의 introduce() 사용 — Dog는 오버라이딩하지 않음
print()
````

````python
# Animal, Cat, Dog 클래스가 정의되어 있어야 합니다 (위 셀 먼저 실행)

print("=== isinstance / issubclass ===")
print()

# isinstance: 객체가 특정 클래스(또는 부모 클래스)의 인스턴스인지 확인
print(f"Is luna a Cat?     {isinstance(luna, Cat)}")
print(f"Is luna an Animal? {isinstance(luna, Animal)}")   # True — Cat은 Animal이다
print(f"Is luna a Dog?     {isinstance(luna, Dog)}")
print()
````

## 4. 캡슐화

### 왜 캡슐화가 필요한가?

잔액(`balance`)을 공개 변수로 두면 어떤 일이 생길까?

```python
account.balance = -999_999   # 음수 잔액? 가능!
account.balance = 9_999_999  # 입금도 없이 부풀리기? 가능!
```

외부 코드가 잔액을 규칙 없이 직접 바꿀 수 있다면,  
`deposit()` / `withdraw()` 에 만들어 둔 검증 로직이 전혀 의미가 없어진다.

→ **캡슐화**는 내부 데이터를 보호하고, 정해진 메서드(`deposit`, `withdraw`)를 통해서만  
데이터를 변경하도록 강제해 **의도하지 않은 상태 변경을 막는다.**

### 네이밍 규칙

| 네이밍 | 의미 | 외부 접근 |
|--------|------|----------|
| `변수` | 공개 (public) | 자유롭게 접근 가능 |
| `_변수` | 보호 (protected) | 관례적 경고 — 접근 자제 요청, 강제는 아님 |
| `__변수` | 비공개 (private) | 이름 맹글링 적용 — 외부에서 `_클래스명__변수`로만 접근 가능 |

### @property

`@property` 데코레이터를 사용하면 메서드를 속성처럼 읽을 수 있다.  
게터(getter)를 구현해 외부에 읽기 전용 접근을 허용할 때 주로 사용한다.

```python
@property
def 잔액(self):       # 계좌.잔액 처럼 () 없이 접근
    return self.__잔액
```

````python
# ── Before: 잔액을 보호하지 않으면? ──────────────────────────────────────
class UnsafeBankAccount:
    def __init__(self, owner, balance=0):
        self.owner   = owner
        self.balance = balance   # 완전 공개 — 누구나 직접 수정 가능

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount

account = UnsafeBankAccount("Alice", 100_000)

# 입금·출금 메서드를 거치지 않고 잔액을 직접 수정할 수 있다
account.balance = -999_999   # 음수 잔액! — deposit/withdraw 검증이 무력화됨
print(f"직접 수정 후 잔액: {account.balance:,}")

account.balance = 9_999_999  # 입금 없이 임의로 부풀리기
print(f"직접 수정 후 잔액: {account.balance:,}")
print()
print("→ deposit/withdraw 에 만들어둔 검증 로직이 전혀 의미 없어진다.")
print("  다음 셀에서 __balance 로 이 문제를 해결한다.")
````

````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner    = owner
        self.__balance = balance      # __는 이름 맹글링 적용 → _BankAccount__balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be greater than 0")
        self.__balance += amount
        print(f"  Deposited +{amount:,}  →  Balance: {self.__balance:,}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be greater than 0")
        if amount > self.__balance:
            raise ValueError(f"Insufficient funds (current balance: {self.__balance:,})")
        self.__balance -= amount
        print(f"  Withdrew  -{amount:,}  →  Balance: {self.__balance:,}")

    @property
    def balance(self):    # 읽기 전용 접근자 — account.balance() 아닌 account.balance로 접근
        return self.__balance

    def __str__(self):
        return f"[{self.owner}] Balance: {self.__balance:,}"

account = BankAccount("Alice", 100_000)
account.deposit(50_000)
account.withdraw(30_000)
print(account)
print(f"Balance via @property: {account.balance:,}")
print()
````

````python

# 클래스 밖에서 account.__balance를 대입하면 새로운 일반 속성이 생성됨 (이름 맹글링 미적용)
# 실제 비공개 속성 _BankAccount__balance는 그대로 유지됨
try:
    account.__balance = 999_999_999
except AttributeError as e:
    print(f"Direct access failed: {e}")

print(f"Balance via @property: {account.balance:,}")
````

````python

# 검증 확인
try:
    account.withdraw(500_000)
except ValueError as e:
    print(f"Withdrawal failed: {e}")
````

## 5. 다형성과 매직 메서드

### 왜 다형성이 필요한가?

다형성 없이 여러 도형의 넓이를 출력하려면?

```python
def print_area(shape):
    if isinstance(shape, Rectangle):
        print(shape.width * shape.length)
    elif isinstance(shape, Triangle):
        print(0.5 * shape.base * shape.height)
    elif isinstance(shape, CircleShape):
        print(math.pi * shape.radius ** 2)
    # 도형이 추가될 때마다 이 함수도 수정해야 한다 → 유지보수 지옥
```

도형이 5개면 `elif` 5개, 10개면 10개.  
더 나쁜 건, 새 도형을 추가할 때마다 이 함수를 찾아서 수정해야 한다.

→ **다형성**을 쓰면 각 클래스가 자신의 `area()`를 직접 책임진다.  
호출하는 쪽은 `for s in shapes: s.area()` 한 줄로 끝난다 — 도형이 늘어나도 이 코드는 바뀌지 않는다.

### 다형성(Polymorphism)이란?

같은 메서드 이름을 호출해도 클래스마다 다른 동작을 하는 것이다.  
공통 인터페이스를 통해 다양한 객체를 일관되게 처리할 수 있다.

### 매직 메서드(Magic Method)

`__이름__` 형태의 특수 메서드로, 파이썬 내장 연산자나 함수와 연동된다.  
클래스에 직접 정의하면 `+`, `==`, `len()`, `print()` 등의 동작을 커스터마이징할 수 있다.

| 매직 메서드 | 연동 연산자/함수 | 설명 |
|------------|----------------|------|
| `__str__(self)` | `str()`, `print()` | 사람이 읽을 수 있는 문자열 표현 |
| `__repr__(self)` | `repr()` | 개발자용 상세 문자열 표현 |
| `__len__(self)` | `len()` | 길이 반환 (정수 반환 필수) |
| `__eq__(self, other)` | `==` | 동등 비교 |
| `__add__(self, other)` | `+` | 덧셈 연산 |
| `__mul__(self, other)` | `*` | 곱셈 연산 |
| `__lt__(self, other)` | `<` | 크기 비교 |

````python
# ── Before: 다형성 없이 isinstance로 분기하면? ───────────────────────────
import math

# 도형의 로직이 이 함수 안에 전부 집중된다
# 새 도형(Pentagon 등)이 추가될 때마다 이 함수를 찾아서 수정해야 한다
def describe_shape_bad(shape_type, **kwargs):
    if shape_type == "rectangle":
        area = kwargs["width"] * kwargs["length"]
    elif shape_type == "triangle":
        area = 0.5 * kwargs["base"] * kwargs["height"]
    elif shape_type == "circle":
        area = math.pi * kwargs["radius"] ** 2
    else:
        raise ValueError(f"Unknown shape: {shape_type}")
    print(f"  {shape_type}: area = {area:.2f}")

describe_shape_bad("rectangle", width=4, length=5)
describe_shape_bad("triangle",  base=6, height=8)
describe_shape_bad("circle",    radius=3)
print()
print("→ 도형이 10개면 elif 10개, 100개면 100개.")
print("  다음 셀에서 다형성을 쓰면 이 함수 자체가 필요 없어진다.")
````

````python
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area()")

    def describe(self):
        return f"{type(self).__name__}: area = {self.area():.2f}"

class Rectangle(Shape):
    def __init__(self, width, length):
        self.width  = width
        self.length = length

    def area(self):
        return self.width * self.length

class Triangle(Shape):
    def __init__(self, base, height):
        self.base   = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

# 2절에서 정의한 Circle 클래스와 이름 충돌을 피하기 위해 CircleShape로 명명
class CircleShape(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        import math
        return math.pi * self.radius ** 2

# 다형성: 동일한 .describe() 호출이 각 클래스의 .area() 구현을 실행함
shapes = [Rectangle(4, 5), Triangle(6, 8), CircleShape(3)]
for s in shapes:
    print(s.describe())

print()
largest = max(shapes, key=lambda s: s.area())
print(f"Largest shape: {largest.describe()}")
````

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):            # print()와 str() 호출 시 실행
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):           # 개발자용 상세 표현
        return f"Vector(x={self.x}, y={self.y})"

    def __add__(self, other):     # + 연산자 호출 시 실행
        return Vector(self.x + other.x, self.y + other.y)

    def __eq__(self, other):      # == 연산자 호출 시 실행
        return self.x == other.x and self.y == other.y

    def __len__(self):            # len() 호출 시 실행 — 정수를 반환해야 함
        import math
        return int(math.sqrt(self.x**2 + self.y**2))

    def __mul__(self, scalar):    # * 연산자 호출 시 실행 (스칼라 곱)
        return Vector(self.x * scalar, self.y * scalar)

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1: {v1}")               # __str__ 호출
print(f"repr: {repr(v1)}")       # __repr__ 호출
print(f"v1 + v2: {v1 + v2}")    # __add__ 호출
print(f"v1 == v2: {v1 == v2}")  # __eq__ 호출
print(f"len(v1): {len(v1)}")    # __len__ 호출 — 크기 5 (3-4-5 직각삼각형)
print(f"v1 * 3: {v1 * 3}")      # __mul__ 호출
````

````python
# 종합 실습: 상속 + 캡슐화 + 다형성 + 매직 메서드
import random

class Character:    # 부모 클래스
    def __init__(self, name, hp, attack_power):
        self.name         = name
        self.__hp         = hp           # 캡슐화 — @property로 접근
        self.__max_hp     = hp
        self.attack_power = attack_power

    @property
    def hp(self):
        return self.__hp

    @property
    def is_alive(self):          # @property — character.is_alive처럼 괄호 없이 접근
        return self.__hp > 0

    def take_damage(self, amount):
        actual = max(0, amount)
        self.__hp = max(0, self.__hp - actual)
        return actual

    def attack(self, target):    # 자식 클래스에서 오버라이딩 예정 (다형성)
        damage = random.randint(int(self.attack_power * 0.8), int(self.attack_power * 1.2))
        actual = target.take_damage(damage)
        return f"{self.name} → {target.name}: {actual} damage!"

    def status_bar(self):
        ratio  = self.__hp / self.__max_hp
        filled = int(ratio * 10)
        return f"[{'■' * filled}{'□' * (10 - filled)}] {self.__hp}/{self.__max_hp}"

    def __str__(self):
        return f"{self.name} HP:{self.status_bar()}"

    def __repr__(self):
        return f"Character(name='{self.name}', hp={self.__hp})"


class Warrior(Character):    # 자식 클래스 — 방어력 스탯 추가
    def __init__(self, name):
        super().__init__(name, hp=120, attack_power=15)   # 부모 __init__ 먼저 호출
        self.defense = 5

    def take_damage(self, amount):    # 오버라이딩 — 방어력만큼 피해 감소
        reduced = max(0, amount - self.defense)
        return super().take_damage(reduced)

    def attack(self, target):
        result = super().attack(target)
        return f"[Sword Strike] {result}"


class Mage(Character):    # 자식 클래스 — 마나 자원 추가
    def __init__(self, name):
        super().__init__(name, hp=80, attack_power=25)    # 부모 __init__ 먼저 호출
        self.__mana = 100

    def attack(self, target):
        if self.__mana >= 20:
            self.__mana -= 20
            damage = int(self.attack_power * 1.5)
            actual = target.take_damage(damage)
            return f"[Magic Spell] {self.name} → {target.name}: {actual} damage! (Mana: {self.__mana})"
        else:
            return f"[Magic Spell] {self.name}: Not enough mana! (Mana: {self.__mana})"


print("Character classes defined successfully.")
print(repr(Warrior("Test")))
````

````python
# 이 셀 실행 전 위의 Character 클래스 정의 셀을 먼저 실행하세요

random.seed(42)

hero  = Warrior("Arthur")
mage1 = Mage("Merlin")
boss  = Warrior("Dark Lord")

print("=== RPG Battle Simulation ===")
print(f"Allied: {hero}")
print(f"Allied: {mage1}")
print(f"Enemy:  {boss}")
print()

allied_team = [hero, mage1]
turn = 1
while boss.is_alive and any(c.is_alive for c in allied_team):
    print(f"--- Round {turn} ---")
    for character in allied_team:
        if character.is_alive and boss.is_alive:
            print(f"  {character.attack(boss)}")
    if boss.is_alive:
        living_allies = [c for c in allied_team if c.is_alive]
        if living_allies:
            target = random.choice(living_allies)
            print(f"  {boss.attack(target)}")
    print(f"  Boss status: {boss}")
    for c in allied_team:
        print(f"  {c}")
    print()
    turn += 1
    if turn > 6:   # 최대 6라운드
        break

if not boss.is_alive:
    print("Allies win!")
elif not any(c.is_alive for c in allied_team):
    print("Boss wins...")
else:
    print("Battle ended (time limit reached)")

# 최종 isinstance 확인
print()
print(f"isinstance(hero, Character):  {isinstance(hero, Character)}")
print(f"isinstance(mage1, Warrior):   {isinstance(mage1, Warrior)}")
````

## 챕터 11 정리

### 이번 챕터에서 배운 내용

- **클래스와 인스턴스**: 설계도(클래스)로 실제 객체(인스턴스)를 만들고, `__init__`과 `self`로 초기화
- **클래스 변수 vs 인스턴스 변수**: 공유 데이터와 개별 데이터를 구분
- **메서드 3종류**: 인스턴스 메서드(`self`), 클래스 메서드(`@classmethod`, `cls`), 정적 메서드(`@staticmethod`)
- **상속**: `super()`로 부모 기능을 계승하고, 오버라이딩으로 자식만의 동작 정의
- **isinstance / issubclass**: 객체와 클래스 간 관계 확인
- **캡슐화**: `__변수`로 데이터 보호, `@property`로 읽기 전용 접근자 구현
- **다형성**: 같은 메서드명으로 클래스마다 다른 동작
- **매직 메서드**: `__str__`, `__repr__`, `__len__`, `__eq__`, `__add__` 등으로 연산자/함수와 연동

### 메서드 3종류 비교

| 종류 | 데코레이터 | 첫 매개변수 | 주요 용도 |
|------|-----------|------------|-----------|
| 인스턴스 메서드 | 없음 | `self` | 인스턴스 데이터 읽기/쓰기 |
| 클래스 메서드 | `@classmethod` | `cls` | 대체 생성자, 클래스 수준 기능 |
| 정적 메서드 | `@staticmethod` | 없음 | 클래스와 관련된 독립 유틸리티 |

### OOP 4대 핵심 원칙

| 원칙 | 설명 | 파이썬 구현 |
|------|------|------------|
| 캡슐화 (Encapsulation) | 데이터와 메서드를 하나로 묶고 내부를 보호 | `__변수`, `@property` |
| 상속 (Inheritance) | 기존 클래스를 확장해 코드 재사용 | `class 자식(부모)`, `super()` |
| 다형성 (Polymorphism) | 같은 인터페이스로 다양한 동작 | 메서드 오버라이딩 |
| 추상화 (Abstraction) | 복잡한 내부를 숨기고 필요한 인터페이스만 노출 | `raise NotImplementedError` |
